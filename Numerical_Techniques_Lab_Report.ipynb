# Implementation of Bisection Method and Fixed-Point Iteration Method




## 1. Objective

The objective of this computational analysis



is to analyze, implement, and compare two fundamental numerical methods for finding the roots of non-linear equations $f(x) = 0$: the **Bisection Method** (a bracketing method) and the **Fixed-Point Iteration Method** (an open method). This comprehensive study examines their mathematical foundations, convergence characteristics, algorithmic frameworks, and practical applications.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

* **Bisection Method:** Rooted in the **Intermediate Value Theorem (IVT)**. If a continuous function $f(x)$ changes sign over an interval $[a, b]$ (i.e., $f(a) \cdot f(b) < 0$), there exists at least one real root $\alpha \in (a, b)$ such that $f(\alpha) = 0$. By systematically halving the interval, the method brackets and traps the root.
* **Fixed-Point Iteration Method:** Transforms the root-finding problem $f(x) = 0$ into an algebraically equivalent fixed-point problem $x = g(x)$. A point $\alpha$ satisfying this condition is called a fixed point of $g(x)$. The method generates a sequence of approximations by repeatedly evaluating $x_{k+1} = g(x_k)$.

### Governing Formulae

For the **Bisection Method**, the midpoint $c_k$ at the $k$-th iteration is defined as:


$$c_k = \frac{a_k + b_k}{2}$$


The next sub-interval $[a_{k+1}, b_{k+1}]$ is determined by checking the sign of the function evaluations:


$$\text{If } f(a_k) \cdot f(c_k) < 0, \text{ then } a_{k+1} = a_k, \, b_{k+1} = c_k$$

$$\text{If } f(c_k) \cdot f(b_k) < 0, \text{ then } a_{k+1} = c_k, \, b_{k+1} = b_k$$

For the **Fixed-Point Iteration Method**, the recursive functional relationship is:


$$x_{k+1} = g(x_k), \quad k = 0, 1, 2, \dots$$

### Convergence Criteria

* **Bisection Method:** Convergence is **guaranteed** provided that $f(x)$ is continuous on $[a, b]$ and $f(a) \cdot f(b) < 0$.
* **Fixed-Point Iteration Method:** Local convergence is guaranteed by the **Banach Fixed-Point Theorem** (or Contraction Mapping Theorem). If $g(x)$ and $g'(x)$ are continuous on an interval $I$ containing the fixed point $\alpha$, the sequence will converge to $\alpha$ from a starting point $x_0 \in I$ if:

$$|g'(x)| \le L < 1 \quad \forall x \in I$$



where $L$ is the Lipschitz contraction constant. If $|g'(\alpha)| > 1$, the sequence diverges.

### Order of Convergence

* **Bisection Method:** Exhibits **linear convergence** with an asymptotic error constant of $\mu = 0.5$. Each iteration guarantees that the absolute error bounds are strictly halved:

$$e_{k+1} = \frac{1}{2} e_k$$


* **Fixed-Point Iteration Method:** Typically exhibits **linear convergence** ($p=1$). The error updates according to:

$$e_{k+1} \approx |g'(\alpha)| e_k$$



*Note: If $g'(\alpha) = 0$, the method can achieve quadratic ($p=2$) or higher-order convergence, depending on the higher-order derivatives.*






## 3. Algorithmic Framework

### Bisection Method Algorithm

1. **Input:** Define continuous function $f(x)$, initial boundaries $a$ and $b$, absolute error tolerance $\epsilon$, and maximum iterations $N_{max}$.
2. **Verify Interval:** Compute $f(a)$ and $f(b)$. If $f(a) \cdot f(b) \ge 0$, terminate with an error stating that the interval does not guarantee a bracketed root.
3. **Initialize:** Set iteration counter $k = 1$.
4. **Compute Midpoint:** Calculate $c_k = a + \frac{b - a}{2}$.
5. **Check Tolerance & Convergence:** If $|f(c_k)| < \epsilon$ or $\frac{b - a}{2} < \epsilon$, stop and output $c_k$ as the root.
6. **Update Interval:**
* If $f(a) \cdot f(c_k) < 0$, set $b = c_k$.
* Else, set $a = c_k$.


7. **Loop Control:** If $k < N_{max}$, increment $k = k + 1$ and return to Step 4. Otherwise, stop and signal that the maximum iterations were reached.

### Fixed-Point Iteration Method Algorithm

1. **Input:** Define function $g(x)$, initial guess $x_0$, absolute error tolerance $\epsilon$, and maximum iterations $N_{max}$.
2. **Initialize:** Set iteration counter $k = 0$.
3. **Compute Next Iteration:** Evaluate $x_{k+1} = g(x_k)$.
4. **Check Convergence:** If $|x_{k+1} - x_k| < \epsilon$ or $|f(x_{k+1})| < \epsilon$, stop and output $x_{k+1}$ as the root.
5. **Loop Control:** If $k < N_{max}$, update $x_0 = x_{k+1}$, increment $k = k + 1$, and return to Step 3. Otherwise, terminate and signal divergence or execution limits.




## 5. Mathematical Verification Case Study

Let us find the real root of the non-linear equation:


$$f(x) = x^3 - x - 1 = 0$$

### Case A: Bisection Method Verification

We establish the initial bracket. Let $a_1 = 1$ and $b_1 = 2$.

* $f(1) = 1^3 - 1 - 1 = -1$
* $f(2) = 2^3 - 2 - 1 = 5$
Since $f(1) \cdot f(2) = -5 < 0$, a unique root exists within $(1, 2)$.
* **Iteration 1:** $c_1 = \frac{1 + 2}{2} = 1.5$.
$f(1.5) = (1.5)^3 - 1.5 - 1 = 0.875$.
Since $f(1) \cdot f(1.5) = (-1)(0.875) < 0$, the root updates to interval $[1, 1.5]$. New $b_2 = 1.5$.
* **Iteration 2:** $c_2 = \frac{1 + 1.5}{2} = 1.25$.
$f(1.25) = (1.25)^3 - 1.25 - 1 = -0.296875$.
Since $f(1.25) \cdot f(1.5) < 0$, the root updates to interval $[1.25, 1.5]$. New $a_3 = 1.25$.
* **Iteration 3:** $c_3 = \frac{1.25 + 1.5}{2} = 1.375$.
Continuing this deterministic framework converges onto the true root $\alpha \approx 1.3247$ to arbitrary tolerance limits.

### Case B: Fixed-Point Iteration Method Verification

Rewrite $x^3 - x - 1 = 0$ as $x = g(x)$. Let's select two candidates for $g(x)$ to evaluate convergence characteristics near the estimated root $\alpha \approx 1.325$.

1. **Candidate 1:** $g_1(x) = x^3 - 1$

$$g_1'(x) = 3x^2 \implies g_1'(1.325) \approx 3(1.325)^2 \approx 5.27$$



Since $|g_1'(x)| > 1$, this choice will **diverge**.
2. **Candidate 2:** $g_2(x) = \sqrt[3]{x + 1} = (x + 1)^{1/3}$

$$g_2'(x) = \frac{1}{3}(x + 1)^{-2/3} \implies g_2'(1.325) = \frac{1}{3}(2.325)^{-2/3} \approx 0.189$$



Since $|g_2'(x)| \approx 0.189 < 1$, this form will **converge unconditionally**.

Executing the sequence with $x_0 = 1.5$ using $g_2(x)$:

* $x_1 = g_2(1.5) = (2.5)^{1/3} \approx 1.3572$
* $x_2 = g_2(1.3572) = (2.3572)^{1/3} \approx 1.3309$
* $x_3 = g_2(1.3309) = (2.3309)^{1/3} \approx 1.3259$

### Error & Convergence Analysis

For the **Bisection Method**, the maximum absolute error bounded at step $k$ is perfectly structured by:


$$E_k \le \frac{b - a}{2^k}$$

$$k > \frac{\ln(b - a) - \ln(\epsilon)}{\ln(2)}$$

$$e_{k+1} \approx |g'(\alpha)| e_k$$


Smaller magnitudes of $|g'(\alpha)|$ correspond to steeper error decay and faster processing execution rates.

## 7. Real-World Applications

* **Engineering Design Optimizations:** Finding structural dimensions where load deflection variations balance to zero.
* **Financial Mathematics:** Calculating internal rates of return (IRR) by identifying the roots of net present value cash-flow expressions.

## 8. Advantages & Limitations

### Bisection Method

* **Advantages:** Exceptionally robust; it always converges if a valid sign change is bounded. It does not require computing complex analytical derivatives.
* **Limitations:** Relatively slow convergence rate. It cannot find roots that touch but do not cross the x-axis (e.g., $f(x) = x^2$), and it cannot handle complex numbers.

### Fixed-Point Iteration Method

* **Advantages:** Requires only one initial guess. The algebraic setup can be highly adaptable, and it scales effectively to systems of multi-variable non-linear equations.
* **Limitations:** Highly sensitive to the choice of $g(x)$. Determining a convergent path formulation where $|g'(x)| < 1$ can be mathematically challenging.

This study demonstrates that root-finding algorithms involve a trade-off between **robustness** and **efficiency**. The Bisection Method is slow but completely reliable, making it an excellent fallback strategy. In contrast, the Fixed-Point Iteration Method offers structural flexibility and faster convergence, but only when it meets strict analytical conditions. In practical applications, these techniques are often combined into hybrid frameworks using the Bisection Method to safely narrow down an interval before switching to a faster open method to compute the precise root quickly.

# Implementation of Regula–Falsi (False Position) Method and Secant Method



## 1. Objective

The objective of this computational analysis is to analyze, implement, and compare the **Regula–Falsi (False Position) Method** and the **Secant Method** for solving non-linear algebraic and transcendental equations of the form $f(x) = 0$. This study evaluates their mathematical principles, geometric interpretations, algorithmic steps, convergence rates, and trade-offs.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

Both methods approximate the root of a function by replacing the non-linear curve $f(x)$ with a linear secant line through two points.

* **Regula–Falsi Method:** A **bracketing method** that utilizes the Intermediate Value Theorem. It requires two initial points, $a$ and $b$, such that $f(a) \cdot f(b) < 0$. The root is approximated by the intersection of the secant line passing through $(a, f(a))$ and $(b, f(b))$ with the x-axis. It updates the interval to always keep the root bracketed.
* **Secant Method:** An **open method** that uses the exact same secant line formula as Regula–Falsi but removes the bracketing constraint. It simply uses the two most recent approximations ($x_k$ and $x_{k-1}$) to project the next point $x_{k+1}$, regardless of whether they bracket a sign change.



### Governing Formulae

The equation of the secant line passing through $(a, f(a))$ and $(b, f(b))$ is given by:


$$y - f(b) = \frac{f(b) - f(a)}{b - a}(x - b)$$

Setting $y = 0$ yields the x-intercept, which serves as the next root approximation.

For the **Regula–Falsi Method**, the next estimate $c_k$ within the bracket $[a_k, b_k]$ is calculated as:


$$c_k = b_k - \frac{f(b_k)(b_k - a_k)}{f(b_k) - f(a_k)} = \frac{a_k f(b_k) - b_k f(a_k)}{f(b_k) - f(a_k)}$$

For the **Secant Method**, using successive iterations $x_{k-1}$ and $x_k$, the sequence progresses via:


$$x_{k+1} = x_k - \frac{f(x_k)(x_k - x_{k-1})}{f(x_k) - f(x_{k-1})} = \frac{x_{k-1}f(x_k) - x_k f(x_{k-1})}{f(x_k) - f(x_{k-1})}$$

### Convergence Criteria

* **Regula–Falsi Method:** Convergence is **guaranteed** if $f(x)$ is continuous on $[a, b]$ and $f(a) \cdot f(b) < 0$. Because it maintains a bracket, it cannot diverge.
* **Secant Method:** Convergence is **conditional**. It requires the initial guesses $x_0$ and $x_1$ to be sufficiently close to the true root $\alpha$. If $f'(x) \approx 0$ during an iteration or if the local geometry causes the secant line to project far outside the root domain, the method can diverge completely.

### Order of Convergence

* **Regula–Falsi Method:** Exhibits **linear convergence** ($p = 1$). While usually faster than the Bisection Method initially, one of the endpoints often becomes stationary, causing the interval to shrink only on one side and slowing down asymptotic convergence.
* **Secant Method:** Exhibits **superlinear convergence** with an order of:

$$p = \frac{1 + \sqrt{5}}{2} \approx 1.618$$



This ratio is known as the **Golden Ratio**. It converges significantly faster than both Bisection and Regula-Falsi without requiring evaluation of the function's analytical derivative.



## 3. Algorithmic Framework

### Regula–Falsi Method Algorithm

1. **Input:** Define the function $f(x)$, initial boundary points $a$ and $b$, error tolerance $\epsilon$, and maximum iterations $N_{max}$.
2. **Verify Bracket:** Check if $f(a) \cdot f(b) \ge 0$. If true, exit with an error (no guaranteed sign change).
3. **Initialize:** Set iteration counter $k = 1$.
4. **Compute Intercept:** Calculate $c_k = \frac{a \cdot f(b) - b \cdot f(a)}{f(b) - f(a)}$.
5. **Check Convergence:** If $|f(c_k)| < \epsilon$ or the change in $c_k$ is less than $\epsilon$, stop and output $c_k$ as the root.
6. **Update Bracket:**
* If $f(a) \cdot f(c_k) < 0$, set $b = c_k$.
* Else, set $a = c_k$.


7. **Loop Control:** If $k < N_{max}$, increment $k = k + 1$ and return to Step 4. Otherwise, terminate.


### Secant Method Algorithm

1. **Input:** Define the function $f(x)$, two initial approximations $x_0$ and $x_1$, error tolerance $\epsilon$, and maximum iterations $N_{max}$.
2. **Initialize:** Set iteration counter $k = 1$.
3. **Check Slope Safeguard:** If $f(x_k) - f(x_{k-1}) \approx 0$, terminate with a division-by-zero/slope failure error.
4. **Compute Next Approximation:** Calculate $x_{k+1} = x_k - \frac{f(x_k)(x_k - x_{k-1})}{f(x_k) - f(x_{k-1})}$.
5. **Check Convergence:** If $|x_{k+1} - x_k| < \epsilon$ or $|f(x_{k+1})| < \epsilon$, stop and output $x_{k+1}$ as the root.
6. **Update Points:** Shift variables: $x_{k-1} = x_k$ and $x_k = x_{k+1}$.
7. **Loop Control:** If $k < N_{max}$, increment $k = k + 1$ and return to Step 3. Otherwise, terminate.




## 5. Mathematical Verification Case Study

Let us locate the root of the equation:


$$f(x) = x^2 - 4x - 10 = 0$$


The exact positive root using the quadratic formula is $\alpha = 2 + \sqrt{14} \approx 5.741657$.

### Case A: Regula–Falsi Method Verification

We choose initial brackets $a_1 = 4$ and $b_1 = 6$.

* $f(4) = 4^2 - 4(4) - 10 = -10$
* $f(6) = 6^2 - 4(6) - 10 = +2$
Since $f(4) \cdot f(6) = -20 < 0$, a root is guaranteed in $[4, 6]$.
* **Iteration 1:**

$$c_1 = \frac{a_1 f(b_1) - b_1 f(a_1)}{f(b_1) - f(a_1)} = \frac{4(2) - 6(-10)}{2 - (-10)} = \frac{8 + 60}{12} = \frac{68}{12} \approx 5.66667$$



Evaluate function at intercept:

$$f(5.66667) = (5.66667)^2 - 4(5.66667) - 10 = -0.55556$$



Since $f(5.66667) \cdot f(6) = (-0.55556)(2) < 0$, the root lies in $[5.66667, 6]$.
Set $a_2 = 5.66667, \, b_2 = 6$.
* **Iteration 2:**

$$c_2 = \frac{5.66667(2) - 6(-0.55556)}{2 - (-0.55556)} = \frac{11.33334 + 3.33336}{2.55556} = \frac{14.6667}{2.55556} \approx 5.73913$$



Notice that the left boundary ($a$) keeps shifting closer to the root while the right boundary ($b=6$) remains frozen as a static pivot point.

### Case B: Secant Method Verification

We choose initial guesses $x_0 = 4$ and $x_1 = 6$.

* $f(x_0) = f(4) = -10$
* $f(x_1) = f(6) = 2$
* **Iteration 1:**

$$x_2 = x_1 - \frac{f(x_1)(x_1 - x_0)}{f(x_1) - f(x_0)} = 6 - \frac{2 \cdot (6 - 4)}{2 - (-10)} = 6 - \frac{4}{12} \approx 5.66667$$



Compute $f(x_2) = f(5.66667) = -0.55556$.
* **Iteration 2:** Now we use the two most recent values: $x_1 = 6$ and $x_2 = 5.66667$.

$$x_3 = x_2 - \frac{f(x_2)(x_2 - x_1)}{f(x_2) - f(x_1)} = 5.66667 - \frac{(-0.55556)(5.66667 - 6)}{-0.55556 - 2}$$


$$x_3 = 5.66667 - \frac{(-0.55556)(-0.33333)}{-2.55556} = 5.66667 - 0.07246 = 5.73921$$



Compute $f(x_3) = f(5.73921) = -0.01777$.
* **Iteration 3:** Shift variables again to use $x_2 = 5.66667$ and $x_3 = 5.73921$.

$$x_4 = 5.73921 - \frac{(-0.01777)(5.73921 - 5.66667)}{-0.01777 - (-0.55556)} \approx 5.74163$$



The Secant Method converges rapidly toward the exact root value of $5.741657$ within fewer total updates due to its superlinear convergence rate.



## 6. Observations & Computational Characteristics

### Error & Convergence Analysis

The error equation for the Secant Method relates errors across three consecutive states:


$$e_{k+1} = C \cdot e_k \cdot e_{k-1}$$


Taking the logarithm of this expression demonstrates that the error satisfies the relationship $e_{k+1} \propto e_k^{1.618}$. This means the number of correct decimal places increases by roughly $61.8\%$ with each iteration, making it significantly faster than the linear convergence of Regula-Falsi.

For Regula-Falsi, if a function maintains its concavity (convex or concave) across the root bracket, one endpoint will get "stuck." As a result, the algorithm effectively behaves like a one-sided linear interpolation, tracking error reductions of $e_{k+1} \approx M \cdot e_k$, where $M$ is tightly bound to the slopes around the stationary node.

### Comparative Analysis

| Feature | Regula–Falsi Method | Secant Method |
| --- | --- | --- |
| **Method Category** | Bracketing / Bounded | Open / Unbounded |
| **Convergence Guarantee** | Guaranteed if sign change exists | Conditional (Can diverge if initial guesses are poor) |
| **Order of Convergence ($p$)** | $1.0$ (Linear) | $\approx 1.618$ (Superlinear / Golden Ratio) |
| **Computational Overhead** | Low (Requires 1 function evaluation per step) | Low (Requires 1 function evaluation per step) |
| **Risk of Failure** | None, but can experience very slow convergence | Division by zero if $f(x_k) = f(x_{k-1})$; oscillating loops |
| **Endpoint Trapping** | Common (One boundary point often gets "stuck") | None (Both endpoints update dynamically) |



## 7. Real-World Applications

* **Chemical Engineering Thermodynamics:** Calculating the specific volume of non-ideal gases using cubic equations of state (e.g., the van der Waals or Redlich-Kwong equations).
* **Structural Engineering Deflection Analysis:** Determining zero-crossing inflection stresses inside complex spatial trusses or beams under asymmetric loading conditions.
* **Aerodynamic Flight Trajectories:** Finding stall angles or boundary-layer transition points where pressure coefficient distributions reach equilibrium thresholds.


## 8. Advantages & Limitations

### Regula–Falsi Method

* **Advantages:** Unconditionally stable and reliable. Unlike the Bisection method, it scales its step size based on the function's values rather than blindly cutting intervals in half.
* **Limitations:** Linear asymptotic convergence can slow down drastically when dealing with highly curved functions, as one endpoint often stalls.

### Secant Method

* **Advantages:** Fast convergence ($p \approx 1.618$) that approaches the speed of Newton’s method without requiring analytical derivatives ($f'(x)$). This makes it ideal for black-box or highly complex objective functions.
* **Limitations:** Unstable if initial guesses are poor. It can fail due to division-by-zero errors if the function values at the two points happen to match.



## 9. Conclusion

This comparative study highlights the trade-off between **safety** and **speed** in numerical root-finding. The Regula-Falsi method prioritizes stability by ensuring the root remains trapped within an interval, making it reliable but prone to slowing down when an endpoint stalls. On the other hand, the Secant method swaps safety for speed, using dynamic updates to achieve faster, superlinear convergence ($p \approx 1.618$). In practice, a smart approach is to start with a few steps of a bracketing method to get close to the root, and then switch to the Secant method to lock onto the precise answer quickly.

# Implementation of Newton–Raphson Method and Generalized Newton’s Method



## 1. Objective / Aim

The objective of this report is to analyze, implement, and compare the standard **Newton–Raphson Method** and its extension, the **Generalized Newton’s Method** (also known as the Schröder's method or modified Newton's method for multiple roots), for finding the real roots of non-linear equations. This study focuses on how these methods perform when dealing with simple roots versus roots with higher multiplicity.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

* **Newton–Raphson Method:** An open method based on a first-order Taylor series expansion of $f(x)$ around an initial guess $x_k$. Geometrically, the method approximates the function by its tangent line at $(x_k, f(x_k))$ and projects the next estimate where this tangent line intersects the x-axis. It is designed for finding **simple roots** (where $f(\alpha) = 0$ and $f'(\alpha) \neq 0$).
* **Generalized Newton’s Method:** Standard Newton-Raphson slows down significantly when approaching a **multiple root** (where $f(\alpha) = 0, f'(\alpha) = 0, \dots, f^{(m-1)}(\alpha) = 0$ for a root of multiplicity $m$). The Generalized Newton's method introduces a correction factor based on the root's multiplicity $m$ to restore rapid convergence.

### Governing Formulae

The first-order Taylor expansion around $x_k$ is:


$$f(x) \approx f(x_k) + f'(x_k)(x - x_k)$$


Setting $f(x) = 0$ and solving for $x$ yields the **Newton–Raphson Formula**:


$$x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)}$$

For a root with a known multiplicity $m$, the **Generalized Newton’s Method Formula** introduces a multiplier to adjust the step size:


$$x_{k+1} = x_k - m \frac{f(x_k)}{f'(x_k)}$$

> **Alternative Form (Unknown Multiplicity):** If the multiplicity $m$ is unknown, we can define a new function $u(x) = \frac{f(x)}{f'(x)}$, which shares the exact same roots as $f(x)$ but converts all multiple roots into simple roots. Applying standard Newton-Raphson to $u(x)$ yields:
>
> $$x_{k+1} = x_k - \frac{u(x_k)}{u'(x_k)} = x_k - \frac{f(x_k)f'(x_k)}{[f'(x_k)]^2 - f(x_k)f''(x_k)}$$
>
>

### Convergence Criteria

* **Newton–Raphson Method:** Local convergence is guaranteed if $f(x)$, $f'(x)$, and $f''(x)$ are continuous in an interval surrounding the root $\alpha$, $f'(\alpha) \neq 0$, and the initial guess $x_0$ is chosen sufficiently close to $\alpha$. If $f'(x_k) \approx 0$ (a flat spot or local extremum), the method fails due to division by zero.
* **Generalized Newton’s Method:** Requires the multiplicity $m$ to be correctly identified, or requires computing up to the second derivative $f''(x)$ if using the alternative formula.

### Order of Convergence

* **Newton–Raphson Method:** Exhibits **quadratic convergence** ($p = 2$) for simple roots. The error updates according to:

$$e_{k+1} \approx \frac{f''(\alpha)}{2f'(\alpha)} e_k^2$$



However, if the root has a multiplicity $m > 1$, standard Newton-Raphson drops to **linear convergence** ($p = 1$) with an asymptotic error constant of $\mu = 1 - \frac{1}{m}$.
* **Generalized Newton’s Method:** Restores **quadratic convergence** ($p = 2$) even for multiple roots, saving considerable computational time.



## 3. Algorithmic Framework (Numbered Steps)

### Newton–Raphson Method Algorithm

1. **Input:** Define function $f(x)$, derivative $f'(x)$, initial guess $x_0$, error tolerance $\epsilon$, and maximum iterations $N_{max}$.
2. **Initialize:** Set iteration counter $k = 0$.
3. **Slope Check:** Evaluate $f'(x_k)$. If $|f'(x_k)| < 10^{-12}$ (or a similarly small threshold), terminate with an error: "Derivative too close to zero; method fails."
4. **Compute Next Point:** Calculate $x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)}$.
5. **Check Convergence:** If $|x_{k+1} - x_k| < \epsilon$ or $|f(x_{k+1})| < \epsilon$, stop and output $x_{k+1}$ as the root.
6. **Loop Control:** If $k < N_{max}$, set $x_0 = x_{k+1}$, increment $k = k + 1$, and return to Step 3. Otherwise, terminate.

### Generalized Newton’s Method Algorithm (Known Multiplicity $m$)

1. **Input:** Define $f(x)$, $f'(x)$, initial guess $x_0$, multiplicity $m$, error tolerance $\epsilon$, and maximum iterations $N_{max}$.
2. **Initialize:** Set iteration counter $k = 0$.
3. **Slope Check:** If $|f'(x_k)| < 10^{-12}$, terminate with a derivative failure error.
4. **Compute Next Point:** Calculate $x_{k+1} = x_k - m \frac{f(x_k)}{f'(x_k)}$.
5. **Check Convergence:** If $|x_{k+1} - x_k| < \epsilon$ or $|f(x_{k+1})| < \epsilon$, stop and output $x_{k+1}$ as the root.
6. **Loop Control:** If $k < N_{max}$, set $x_0 = x_{k+1}$, increment $k = k + 1$, and return to Step 3. Otherwise, terminate.

Factoring this equation gives $(x - 1)^2(x + 2) = 0$. Thus, $\alpha = 1$ is a root with a **multiplicity of $m = 2$**.
The derivative functions are:

*Error:* $e_1 = |1.10303 - 1| = 0.10303$ (roughly cut in half).

# Implementation of Forward, Backward, and Central Difference Methods



## 1. Objective

The objective of this report is to analyze, implement, and compare the three primary finite difference approximations for numerical differentiation: the **Forward Difference Method**, the **Backward Difference Method**, and the **Central Difference Method**. This study evaluates their mathematical derivations using Taylor series expansions, their truncation errors, and their computational accuracy.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

Numerical differentiation estimates the derivative of a continuous function $f(x)$ at a specific point $x$ using its values at discrete, equally spaced points. The spacing between these points is defined as the step size $h$.

These methods are derived directly from the **Taylor Series Expansion** of a infinitely differentiable function $f(x)$ around the point $x$:


$$f(x + h) = f(x) + hf'(x) + \frac{h^2}{2!}f''(x) + \frac{h^3}{3!}f'''(x) + \dots$$

$$f(x - h) = f(x) - hf'(x) + \frac{h^2}{2!}f''(x) - \frac{h^3}{3!}f'''(x) + \dots$$

### Governing Formulae

* **Forward Difference Method:** Derived by isolating $f'(x)$ in the expansion for $f(x+h)$ and truncating higher-order terms:

$$f'(x) \approx \frac{f(x + h) - f(x)}{h}$$


* **Backward Difference Method:** Derived by isolating $f'(x)$ in the expansion for $f(x-h)$ and truncating higher-order terms:

$$f'(x) \approx \frac{f(x) - f(x - h)}{h}$$


* **Central Difference Method:** Derived by subtracting the expansion of $f(x-h)$ from the expansion of $f(x+h)$. This cancels out the symmetric even-powered derivative terms ($h^2$):

$$f(x + h) - f(x - h) = 2hf'(x) + \frac{2h^3}{3!}f'''(x) + \dots$$


$$f'(x) \approx \frac{f(x + h) - f(x - h)}{2h}$$



### Convergence & Stability Criteria

As the step size $h$ approaches zero ($h \to 0$), the numerical approximation converges to the exact analytical derivative. However, in practical computing, reducing $h$ indefinitely leads to a trade-off:

* **Truncation Error:** Decreases as $h$ gets smaller.
* **Round-off Error:** Increases as $h$ gets smaller due to the limits of computer precision when subtracting two nearly identical numbers (loss of significance).

Therefore, computational stability requires finding an optimal step size $h_{opt}$ that balances both types of error.

### Order of Convergence (Error Analysis)

The **Big-O notation** ($\mathcal{O}$) indicates how quickly the truncation error shrinks relative to the step size $h$:

* **Forward Difference:** $\mathcal{O}(h)$ (First-order accurate). Cutting $h$ in half reduces the error by roughly half.
* **Backward Difference:** $\mathcal{O}(h)$ (First-order accurate).
* **Central Difference:** $\mathcal{O}(h^2)$ (Second-order accurate). Cutting $h$ in half reduces the error by a factor of **four**, making it significantly more accurate for a given step size.


## 3. Algorithmic Framework (Numbered Steps)

### Finite Difference Estimation Algorithm

1. **Input:** Define the function $f(x)$ (or a discrete dataset), target evaluation point $x$, step size $h$, and selected method choice (`Forward`, `Backward`, or `Central`).
2. **Step Size Check:** Ensure $h > 0$. If $h \le 0$, terminate with an error.
3. **Branch Based on Selected Method:**
* **If `Forward`:** Compute $f(x)$ and $f(x + h)$.
Set $\text{Derivative} = \frac{f(x + h) - f(x)}{h}$.
* **If `Backward`:** Compute $f(x)$ and $f(x - h)$.
Set $\text{Derivative} = \frac{f(x) - f(x - h)}{h}$.
* **If `Central`:** Compute $f(x + h)$ and $f(x - h)$.
Set $\text{Derivative} = \frac{f(x + h) - f(x - h)}{2h}$.


4. **Output:** Return the computed numerical derivative value.



## 5. Mathematical Verification Case Study

Let us approximate the first derivative of the function:


$$f(x) = \ln(x)$$


at the evaluation point $x = 2.0$ using a step size of $h = 0.1$.

The exact analytical derivative is $f'(x) = \frac{1}{x}$. At $x = 2.0$, the **exact value is $0.5$**.

### Functional Evaluations

* $f(x + h) = \ln(2.1) \approx 0.74193734$

### Case B: Backward Difference Approximation

$$f'_{backward}(2.0) \approx \frac{0.69314718 - 0.64185389}{0.1} = \frac{0.05129329}{0.1} = 0.51293290$$

$$\text{Absolute Error} = |0.5 - 0.51293290| = 0.01293290$$

### Case C: Central Difference Approximation

$$f'_{central}(2.0) \approx \frac{0.74193734 - 0.64185389}{2(0.1)} = \frac{0.10008345}{0.2} = 0.50041725$$

## 6. Observations & Computational Characteristics

### Geometric Interpretation

The three methods look at different points on the curve to estimate the slope at a given target point.

* **Forward Difference** uses the target point and a point ahead of it, making it sensitive to changes downstream.
* **Backward Difference** uses the target point and a point behind it, reflecting historical or upstream trends.
* **Central Difference** balances both sides by averaging the slopes, which naturally smooths out asymmetrical bias and yields a much more accurate estimate.

### Comparative Analysis

| Feature | Forward Difference | Backward Difference | Central Difference |
| --- | --- | --- | --- |
| **Points Used** | $x$ and $x+h$ | $x$ and $x-h$ | $x+h$ and $x-h$ |
| **Accuracy Order** | $\mathcal{O}(h)$ (First-order) | $\mathcal{O}(h)$ (First-order) | $\mathcal{O}(h^2)$ (Second-order) |
| **Error Scaling** | Linear decay with $h$ | Linear decay with $h$ | Quadratic decay with $h^2$ |
| **Boundary Suitability** | Excellent at left boundaries | Excellent at right boundaries | Poor at boundaries (requires phantom points) |
| **Computational Cost** | Low (2 function calls) | Low (2 function calls) | Low (2 function calls) |



## 7. Real-World Applications

* **Computational Fluid Dynamics (CFD):** Solving Navier-Stokes transport dynamics on spatial grids.
* **Time-Series Analysis:** Computing velocity, acceleration, or volatile market tickers from discrete real-time data streams.
* **Computer Vision Edge Detection:** Applying Sobel filters or gradient operators to discover sharp boundaries and edges in images.
* **Structural Finite Element Analysis:** Calculating stress distribution maps by solving differential equations for deformed beams.



## 8. Advantages & Limitations

### Forward & Backward Difference Methods

* **Advantages:** Ideal for endpoints or boundary steps where data is only available on one side (e.g., at the start or end of a time series).
* **Limitations:** Lower accuracy ($\mathcal{O}(h)$). Achieving high precision requires choosing a very small step size $h$, which increases the risk of round-off errors.

### Central Difference Method

* **Advantages:** Higher accuracy ($\mathcal{O}(h^2)$) for the exact same number of function evaluations, making it highly computationally efficient.
* **Limitations:** Cannot be used at boundaries without estimating extra "phantom points" outside the valid data range.



## 9. Conclusion

This report illustrates the balance between data availability and accuracy in numerical differentiation. While the Forward and Backward methods are essential for handling boundaries where data is one-sided, they are limited to first-order accuracy. The Central Difference method leverages symmetry to cancel out first-order error terms, providing second-order accuracy ($\mathcal{O}(h^2)$) without requiring extra computation. In practice, high-performance scientific simulations combine these techniques—using central differences across the main domain for accuracy, and switching to forward or backward differences at the boundaries.

# Implementation of Newton’s Forward Interpolation and Newton’s Backward Interpolation



## 1. Objective

The objective of this report is to analyze, implement, and compare **Newton’s Forward Interpolation** and **Newton’s Backward Interpolation** methods. These polynomial interpolation techniques estimate intermediate values of a dependent variable $y = f(x)$ from a set of discrete, equally spaced data points. This study focuses on constructing finite difference tables and choosing the right method based on where the target point lies in the dataset.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

Interpolation involves fitting a polynomial of degree $n$ through $n+1$ known data points $(x_0, y_0), (x_1, y_1), \dots, (x_n, y_n)$ such that the polynomial matches the data points exactly.

When the independent variables ($x$) are equally spaced with a constant step size $h = x_{k+1} - x_k$, we can simplify the interpolation using finite differences:

* **Forward Differences ($\Delta$):** Measure changes moving downward/forward through the dataset, starting from the top.

$$\Delta y_k = y_{k+1} - y_k$$


$$\Delta^2 y_k = \Delta y_{k+1} - \Delta y_k$$


* **Backward Differences ($\nabla$):** Measure changes moving upward/backward through the dataset, starting from the bottom.

$$\nabla y_k = y_k - y_{k-1}$$


$$\nabla^2 y_k = \nabla y_k - \nabla y_{k-1}$$



### Governing Formulae

To simplify the polynomial equations, we define a dimensionless step parameter $u$:

For **Forward Interpolation** (measured relative to the initial point $x_0$):


$$u = \frac{x - x_0}{h}$$


The **Newton’s Forward Difference Interpolation Formula** is:


$$P_n(x) = y_0 + u\Delta y_0 + \frac{u(u-1)}{2!} \Delta^2 y_0 + \frac{u(u-1)(u-2)}{3!} \Delta^3 y_0 + \dots + \frac{u(u-1)\dots(u-n+1)}{n!} \Delta^n y_0$$

For **Backward Interpolation** (measured relative to the final base point $x_n$):


$$u = \frac{x - x_n}{h}$$


The **Newton’s Backward Difference Interpolation Formula** is:


$$P_n(x) = y_n + u\nabla y_n + \frac{u(u+1)}{2!} \nabla^2 y_n + \frac{u(u+1)(u+2)}{3!} \nabla^3 y_n + \dots + \frac{u(u+1)\dots(u+n-1)}{n!} \nabla^n y_n$$

### Convergence & Error Analysis

The interpolation error $E(x) = f(x) - P_n(x)$ for a continuous function with $n+1$ derivatives is bounded by:


$$E(x) = \frac{(x-x_0)(x-x_1)\dots(x-x_n)}{(n+1)!} f^{(n+1)}(\xi)$$


where $\xi$ lies within the interval containing all data points.

* **Convergence Behavior:** Adding more data points generally reduces tracking errors. However, adding too many high-degree terms can cause large oscillations near the edges of the interval—a problem known as **Runge's Phenomenon**.



## 3. Algorithmic Framework (Numbered Steps)

### Finite Difference Table & Interpolation Algorithm

1. **Input:** Read $n+1$ equally spaced data points as arrays $X$ and $Y$, along with the target evaluation value $x_{target}$.
2. **Verify Spacing:** Calculate the step size $h = X[1] - X[0]$. Ensure that $X[k+1] - X[k] = h$ for all steps to verify the data is equally spaced.
3. **Construct Difference Table:** Create a 2D matrix $D$ of size $(n+1) \times (n+1)$. Set the first column $D[i][0] = Y[i]$. Compute higher-order differences column by column:

$$D[i][j] = D[i+1][j-1] - D[i][j-1] \quad \text{for } j = 1 \dots n, \, i = 0 \dots n-j$$


4. **Select Interpolation Strategy:**
* **Forward Path:** If $x_{target}$ is closer to the beginning of the dataset ($x_0$), set $u = \frac{x_{target} - X[0]}{h}$. Initialize $\text{Result} = D[0][0]$. Loop to calculate and add forward terms using the top row of differences ($D[0][j]$).
* **Backward Path:** If $x_{target}$ is closer to the end of the data range ($x_n$), set $u = \frac{x_{target} - X[n]}{h}$. Initialize $\text{Result} = D[n][0]$. Loop to calculate and add backward terms using the bottom-diagonal elements ($D[n-j][j]$).


5. **Output:** Return the computed value as the interpolated result.

# Implementation of Gauss Forward Interpolation and Gauss Backward Interpolation



## 1. Objective

The objective of this report is to analyze, implement, and compare **Gauss Forward Interpolation** and **Gauss Backward Interpolation** methods. These are central difference interpolation techniques designed to estimate intermediate values of a dependent variable $y = f(x)$ from equally spaced data points when the target value lies near the center of the dataset.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

While Newton's forward and backward methods are optimized for values near the boundaries of a dataset, they suffer from increased truncation and round-off errors when applied to central values. Central difference formulas provide higher accuracy for central points by utilizing data points symmetrically distributed on both sides of the target value.

To implement Gauss's central formulas, the standard indexing of data points is shifted so that the origin ($x_0$) is placed at the closest baseline point to the target value. The data points are indexed as:


$$\dots, x_{-3}, x_{-2}, x_{-1}, x_0, x_1, x_2, x_3, \dots$$

### Governing Formulae

Let the constant step size be $h = x_{k+1} - x_k$. We define the dimensionless central step parameter $u$ relative to the chosen origin $x_0$:


$$u = \frac{x - x_0}{h} \quad (\text{where } 0 \le |u| \le 0.5)$$

* **Gauss Forward Interpolation Formula:** This method uses the differences cascades downward and then upward across the central row line. It is optimized for values of $u$ lying between $0$ and $0.5$ ($0 \le u \le 0.5$).

$$P(x) = y_0 + u\Delta y_0 + \frac{u(u-1)}{2!} \Delta^2 y_{-1} + \frac{(u+1)u(u-1)}{3!} \Delta^3 y_{-1} + \frac{(u+1)u(u-1)(u-2)}{4!} \Delta^4 y_{-2} + \dots$$


* **Gauss Backward Interpolation Formula:** This method starts along an upward-trending difference cascade before descending. It is optimized for values of $u$ lying between $-0.5$ and $0$ ($-0.5 \le u \le 0$).

$$P(x) = y_0 + u\Delta y_{-1} + \frac{(u+1)u}{2!} \Delta^2 y_{-1} + \frac{(u+1)u(u-1)}{3!} \Delta^3 y_{-2} + \frac{(u+2)(u+1)u(u-1)}{4!} \Delta^4 y_{-2} + \dots$$



### Convergence Criteria

Like other polynomial interpolation methods, the error converges to zero as $h \to 0$, provided the underlying function is smooth and continuous. Because Gauss methods use differences surrounding the center point symmetrically, they minimize high-order boundary errors and are much less susceptible to **Runge's Phenomenon** across central zones compared to high-degree boundary formulas.



## 3. Algorithmic Framework (Numbered Steps)

### Gauss Central Interpolation Algorithm

1. **Input:** Read $n+1$ equally spaced data points as arrays $X$ and $Y$, and the target point $x_{target}$.
2. **Find Origin index ($i_{origin}$):** Locate the point $X[i]$ that is closest to $x_{target}$. Label this index as $0$ to establish the central coordinate tracking grid.
3. **Compute Step Parameters:** Set $h = X[1] - X[0]$ and calculate $u = \frac{x_{target} - X[i_{origin}]}{h}$.
4. **Construct Central Difference Table:** Build a standard finite difference table where rows are mapped using relative indices $(-k \dots +k)$ centered around $i_{origin}$.
5. **Branch into Polynomial Evaluation:**
* **If $0 \le u \le 0.5$ (Gauss Forward Method):** Select the difference trajectory starting at $y_0$, moving to $\Delta y_0$, then up to $\Delta^2 y_{-1}$, across to $\Delta^3 y_{-1}$, and down to $\Delta^4 y_{-2}$.
* **If $-0.5 \le u < 0$ (Gauss Backward Method):** Select the trajectory starting at $y_0$, moving up to $\Delta y_{-1}$, down to $\Delta^2 y_{-1}$, up to $\Delta^3 y_{-2}$, and down to $\Delta^4 y_{-2}$.


6. **Output:** Return the evaluated polynomial sum as the interpolated result.

# Implementation of Stirling’s Interpolation and Lagrange Interpolation



## 1. Objective
The objective of this exploratory analysis is to analyze, implement, and compare **Stirling’s Interpolation formula** (a central difference method) and **Lagrange Interpolation formula** (an open, globally weighted method). This study highlights how Stirling's method optimizes central calculations for equally spaced points, while Lagrange's method provides a flexible framework that handles unevenly spaced data.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

* **Stirling’s Interpolation Method:** This method takes the average of the Gauss Forward and Gauss Backward interpolation formulas. By combining their geometric tracking paths, it creates a highly balanced central difference formula that cancels out asymmetric error terms. It requires equally spaced data points.
* **Lagrange Interpolation Method:** Unlike finite difference methods, Lagrange interpolation does not require data points to be equally spaced. It constructs a unique polynomial of degree $n$ through $n+1$ points by calculating a weighted sum of **Lagrange fundamental polynomials (weights)** $L_k(x)$, which equal 1 at their corresponding data point $x_k$ and 0 at all other data points.

### Governing Formulae

Let the constant step size be $h$, and the centered step parameter be $u = \frac{x - x_0}{h}$. The **Stirling’s Interpolation Formula** is:


$$P(x) = y_0 + u \left( \frac{\Delta y_0 + \Delta y_{-1}}{2} \right) + \frac{u^2}{2!} \Delta^2 y_{-1} + \frac{u(u^2 - 1)}{3!} \left( \frac{\Delta^3 y_{-1} + \Delta^3 y_{-2}}{2} \right) + \frac{u^2(u^2 - 1)}{4!} \Delta^4 y_{-2} + \dots$$

For a dataset with $n+1$ arbitrary points $(x_0, y_0), (x_1, y_1), \dots, (x_n, y_n)$, the **Lagrange Interpolation Formula** is defined as:


$$P_n(x) = \sum_{k=0}^{n} y_k L_k(x)$$


where the weighting polynomials $L_k(x)$ are calculated as:


$$L_k(x) = \prod_{\substack{j=0 \\ j \neq k}}^{n} \frac{x - x_j}{x_k - x_j} = \frac{(x-x_0)(x-x_1)\dots(x-x_{k-1})(x-x_{k+1})\dots(x-x_n)}{(x_k-x_0)(x_k-x_1)\dots(x_k-x_{k-1})(x_k-x_{k+1})\dots(x_k-x_n)}$$

### Convergence Criteria

* **Stirling's Method:** Local convergence is highly stable and accurate when evaluating values near the center of a dataset ($-0.25 \le u \le 0.25$). It requires the data points to be equally spaced.
* **Lagrange's Method:** Converges to a unique polynomial of degree $\le n$. Because it acts as a global polynomial, it can suffer from severe edge oscillations (**Runge's Phenomenon**) if the polynomial degree gets too high ($n > 5$). It does not require any specific data spacing.


## 3. Algorithmic Framework (Numbered Steps)

### Stirling’s Interpolation Algorithm

1. **Input:** Read $n+1$ equally spaced data points as arrays $X$ and $Y$, and the target evaluation point $x_{target}$.
2. **Find Center Anchor:** Locate the index $i_{origin}$ where $X[i]$ is closest to $x_{target}$. Calculate $h = X[1] - X[0]$ and $u = \frac{x_{target} - X[i_{origin}]}{h}$.
3. **Construct Difference Table:** Build a standard finite difference table.
4. **Compute Polynomial Terms:** Initialize the result sum with $y_0$. Loop through higher-order difference columns:
* For **odd orders** ($1, 3, \dots$), calculate and add the average of the two adjacent central differences multiplied by $\frac{u(u^2-1)\dots}{(order)!}$.
* For **even orders** ($2, 4, \dots$), calculate and add the single direct central difference multiplied by $\frac{u^2(u^2-1)\dots}{(order)!}$.


5. **Output:** Return the final calculated value.

### Lagrange Interpolation Algorithm

1. **Input:** Read $n+1$ arbitrary data points as arrays $X$ and $Y$, along with the target valuation point $x_{target}$.
2. **Initialize:** Set $\text{FinalValue} = 0.0$.
3. **Outer Weight Loop:** Loop through each data point $k$ from $0$ to $n$.
4. **Inner Product Loop:** For a fixed $k$, initialize the term weight $L_k = 1.0$. Loop through all indices $j$ from $0$ to $n$:
* If $j \neq k$, update the weight: $L_k = L_k \cdot \frac{x_{target} - X[j]}{X[k] - X[j]}$.


5. **Accumulate:** Add the weighted value to the running total: $\text{FinalValue} = \text{FinalValue} + (Y[k] \cdot L_k)$.
6. **Loop Control:** Repeat step 4 until all data points are processed, then output $\text{FinalValue}$.

# Implementation of Newton’s Divided Difference Method and Inverse Interpolation



## 1. Objective

The objective of this computational run is to analyze, implement, and compare **Newton’s Divided Difference Method** and the technique of **Inverse Interpolation**. This study focuses on how Newton's divided differences fit unique polynomials to unevenly spaced datasets, and how inverse interpolation reverses this relationship to find the independent value $x$ that corresponds to a given target functional value $y$.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

* **Newton’s Divided Difference Method:** Designed for **unevenly spaced data points**. It introduces the concept of divided differences, which measure the change in a function relative to the varying intervals between its data points. These differences serve as the coefficients for a dynamic polynomial that can easily scale up when new data is added.
* **Inverse Interpolation:** Used when we are given a target value $y_t$ and want to find the corresponding independent value $x_t$ such that $f(x_t) = y_t$. This can be solved using two primary approaches:
1. **Lagrange Inverse Interpolation:** Reverses the variables in the dataset, treating $y$ as the independent variable and $x$ as the dependent variable.
2. **Iterative Root-Finding:** Fits a standard forward polynomial $P_n(x)$ to the original data, then uses an iterative method (like Successive Approximation or Newton-Raphson) to solve the equation $P_n(x) - y_t = 0$.



### Governing Formulae

For a collection of arbitrary data points $(x_0, y_0), (x_1, y_1), \dots, (x_n, y_n)$, the **Divided Differences** are defined recursively:

* First-order:

$$f[x_0, x_1] = \frac{y_1 - y_0}{x_1 - x_0}$$


* Second-order:

$$f[x_0, x_1, x_2] = \frac{f[x_1, x_2] - f[x_0, x_1]}{x_2 - x_0}$$


* Higher-order $k$:

$$f[x_0, x_1, \dots, x_k] = \frac{f[x_1, x_2, \dots, x_k] - f[x_0, x_1, \dots, x_{k-1}]}{x_k - x_0}$$



The **Newton’s Divided Difference Interpolation Formula** uses these coefficients to construct the polynomial:


$$P_n(x) = y_0 + (x - x_0)f[x_0, x_1] + (x - x_0)(x - x_1)f[x_0, x_1, x_2] + \dots + \prod_{j=0}^{n-1}(x - x_j) f[x_0, \dots, x_n]$$

For **Inverse Interpolation via Lagrange’s Method**, we swap the roles of $x$ and $y$:


$$x = \sum_{k=0}^{n} x_k \prod_{\substack{j=0 \\ j \neq k}}^{n} \frac{y - y_j}{y_k - y_j}$$

### Convergence Criteria

* **Divided Differences:** Converges to a unique interpolation polynomial of degree $\le n$. Like Lagrange interpolation, using very high-degree polynomials over wide spans can cause large errors near the boundaries due to **Runge's Phenomenon**.
* **Inverse Interpolation:** Direct variable swapping requires the underlying function to be strictly **monotonic** (either always increasing or always decreasing) across the dataset. If the function changes direction, swapping the variables creates a multi-valued relationship, causing the method to fail.



## 3. Algorithmic Framework (Numbered Steps)

### Newton’s Divided Difference Algorithm

1. **Input:** Read arrays $X$ and $Y$ containing $n+1$ arbitrary data points, along with the target valuation point $x_{target}$.
2. **Construct Table:** Initialize a 2D matrix $DD$ of size $(n+1) \times (n+1)$. Fill the first column with $Y$: $DD[i][0] = Y[i]$.
3. **Compute Differences:** Calculate higher-order columns using nested loops:

$$DD[i][j] = \frac{DD[i+1][j-1] - DD[i][j-1]}{X[i+j] - X[i]} \quad \text{for } j = 1 \dots n, \, i = 0 \dots n-j$$


4. **Evaluate Polynomial:** Initialize $\text{Result} = DD[0][0]$ and a tracking product variable $\text{Term} = 1.0$. Loop through columns $j$ from $1$ to $n$:
* Update the tracking term: $\text{Term} = \text{Term} \cdot (x_{target} - X[j-1])$.
* Accumulate the result: $\text{Result} = \text{Result} + (DD[0][j] \cdot \text{Term})$.


5. **Output:** Return the calculated $\text{Result}$ value.

### Lagrange Inverse Interpolation Algorithm

1. **Input:** Read arrays $X$ and $Y$ of size $n+1$, along with the target functional value $y_{target}$.
2. **Initialize:** Set final variable estimate $x_{interpolated} = 0.0$.
3. **Outer Loop:** Loop through each data index $k$ from $0$ to $n$.
4. **Inner Loop (Swapped Weight):** Initialize the inverse weighting factor $W_k = 1.0$. For every index $j$ from $0$ to $n$:
* If $j \neq k$, update the weight using $y$ values: $W_k = W_k \cdot \frac{y_{target} - Y[j]}{Y[k] - Y[j]}$.


5. **Accumulate:** Update the running total using $x$ values: $x_{interpolated} = x_{interpolated} + (X[k] \cdot W_k)$.
6. **Loop Control:** Repeat until all points are processed, then output $x_{interpolated}$.

# Curve Fitting using Power Curve and Fitting using Exponential Curve



## 1. Objective

The objective of this report is to analyze, implement, and compare non-linear regression models for curve fitting—specifically the **Power Curve** ($y = ax^b$) and the **Exponential Curve** ($y = ae^{bx}$). This study focuses on using logarithmic transformations to linearize these equations, allowing the application of the linear least-squares method to determine the optimal parameters for a given dataset.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

Curve fitting finds a continuous functional relationship that best represents a trend across discrete data points. When data exhibits non-linear growth or decay, linear regression fails. Instead of solving complex non-linear optimization systems directly, we apply mathematical transformations (logarithms) to convert power and exponential relationships into equivalent linear systems:

* **Least-Squares Regression Principle:** Minimizes the sum of the squared residuals (the vertical distances between the observed data points and the fitted regression line).

### Governing Formulae

#### Case A: Exponential Curve ($y = ae^{bx}$)

To linearize the exponential model, take the natural logarithm ($\ln$) of both sides:


$$\ln(y) = \ln(a) + bx$$


Let $Y = \ln(y)$, $A = \ln(a)$, and $B = b$. This yields the standard linear form:


$$Y = A + Bx$$

The normal equations to solve for the parameters $A$ and $B$ are:


$$n A + B \sum x = \sum Y$$

$$A \sum x + B \sum x^2 = \sum (xY)$$

Once $A$ and $B$ are found, the original coefficient $a$ is recovered via:


$$a = e^A$$

#### Case B: Power Curve ($y = ax^b$)

To linearize the power model, take the natural logarithm ($\ln$) of both sides:


$$\ln(y) = \ln(a) + b\ln(x)$$


Let $Y = \ln(y)$, $X = \ln(x)$, and $A = \ln(a)$. This yields the linear form:


$$Y = A + bX$$

The normal equations to solve for the parameters $A$ and $b$ are:


$$n A + b \sum X = \sum Y$$

$$A \sum X + b \sum X^2 = \sum (XY)$$

The original coefficient $a$ is recovered via:


$$a = e^A$$

### Convergence & Optimization Criteria

Unlike iterative root-finding methods, transformed linear least-squares regression does not require checking for convergence loops. It solves for the parameters directly using deterministic matrix operations.

The quality of the fit is evaluated using the **Coefficient of Determination ($R^2$)**:


$$R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$


where $\hat{y}_i$ represents the predicted values and $\bar{y}$ is the mean of the observed data. An $R^2$ value close to $1.0$ indicates an excellent fit.



## 3. Algorithmic Framework (Numbered Steps)

### Transformed Curve Fitting Algorithm

1. **Input:** Read $n$ data points as coordinate arrays $x$ and $y$, and choose the desired model type (`Exponential` or `Power`).
2. **Data Validation:** Ensure all dependent values $y_i > 0$ (and $x_i > 0$ for the Power model) because logarithms cannot process zero or negative numbers.
3. **Apply Transformations:**
* **If `Exponential`:** Compute $Y_i = \ln(y_i)$ and keep the original $x_i$.
* **If `Power`:** Compute $X_i = \ln(x_i)$ and $Y_i = \ln(y_i)$.


4. **Accumulate Sums:** Compute the statistical summary values: $\sum x$ (or $\sum X$), $\sum Y$, $\sum x^2$ (or $\sum X^2$), and $\sum (xY)$ (or $\sum (XY)$).
5. **Solve Normal Equations:** Use Cramer's rule or Gaussian elimination to solve the $2 \times 2$ linear system for the intercept $A$ and slope $B$ (or $b$).
6. **Map Parameters Back:** Recover the original scale coefficient: $a = e^A$.
7. **Output:** Return the calculated parameters $a$ and $b$ along with the structural equation string.


## 5. Mathematical Verification Case Study

Consider the following experimental dataset:

| $x$ | $y$ |
| --- | --- |
| **1** | 2.4 |
| **2** | 5.9 |
| **3** | 14.7 |
| **4** | 36.3 |

We will fit an **Exponential Curve** ($y = ae^{bx}$) to this data.

### Step 1: Compute Transformations and Sums ($Y = \ln(y)$)

| $x$ | $y$ | $Y = \ln(y)$ | $x^2$ | $x \cdot Y$ |
| --- | --- | --- | --- | --- |
| 1 | 2.4 | 0.8755 | 1 | 0.8755 |
| 2 | 5.9 | 1.7750 | 4 | 3.5500 |
| 3 | 14.7 | 2.6879 | 9 | 8.0637 |
| 4 | 36.3 | 3.5918 | 16 | 14.3672 |
| **$\sum x = 10$** |  | **$\sum Y = 8.9302$** | **$\sum x^2 = 30$** | **$\sum xY = 26.8564$** |

Here, the number of data points is $n = 4$.

### Step 2: Set Up Normal Equations

Substitute the sums into the normal equations system:


$$4A + 10B = 8.9302$$

$$10A + 30B = 26.8564$$

### Step 3: Solve for $A$ and $B$

Multiply the first equation by $2.5$:


$$10A + 25B = 22.3255$$

Subtract this from the second equation to isolate $B$:


$$(30 - 25)B = 26.8564 - 22.3255$$

$$5B = 4.5309 \implies B = \mathbf{0.90618}$$

Substitute $B$ back into the first equation to find $A$:


$$4A + 10(0.90618) = 8.9302$$

$$4A + 9.0618 = 8.9302 \implies 4A = -0.1316 \implies A = \mathbf{-0.0329}$$

### Step 4: Map Parameters Back to Original Scale

$$a = e^A = e^{-0.0329} \approx \mathbf{0.9676}$$

$$b = B = \mathbf{0.9062}$$

The final fitted exponential curve equation is:


$$\mathbf{y = 0.9676 \cdot e^{0.9062x}}$$



## 6. Observations & Computational Characteristics

### Geometric Profile Transformations

Linearizing these curves changes how the data is scaled and visualized.

* **Exponential Curves** become a straight line when plotted on a semi-log scale (where only the y-axis uses a logarithmic scale). This model is ideal for data showing rapid growth or decay, such as bacterial growth or radioactive decay, where changes are proportional to the current amount.
* **Power Curves** become a straight line when plotted on a log-log scale (where both the x and y axes use logarithmic scales). This model is ideal for physics laws or geometric relationships where changes scale according to relative proportions.

### Comparative Analysis

| Feature | Exponential Curve Model | Power Curve Model |
| --- | --- | --- |
| **Mathematical Equation** | $y = a \cdot e^{bx}$ | $y = a \cdot x^b$ |

| **Coordinate Adjustments** | Fits $x$ versus $\ln(y)$ (Semi-Log) | Fits $\ln(x)$ versus $\ln(y)$ (Log-Log) |
| **Data Constraints** | Requires $y_i > 0$ | Requires both $x_i > 0$ and $y_i > 0$ |
| **Primary Use Cases** | Compound interest, radioactive decay, population growth trends | Scaling laws, stress-strain curves, fluid turbulence |
| **Growth Characteristics** | Rate of growth accelerates over time | Growth scales relative to size |



## 7. Real-World Applications

* **Biomedical Engineering Kinetics:** Modeling how drug concentrations decay in the bloodstream over time using exponential models.
* **Astrophysics Scaling Laws:** Modeling planetary orbital periods relative to their distance from a star (Kepler's Third Law) using power curves.
* **Material Strength Analysis:** Fitting fatigue life curves (S-N stress curves) to predict when structural components will fail under repeated loads.
* **Epidemiology Modeling:** Tracking how a virus spreads through a population during the early stages of an outbreak.



## 8. Advantages & Limitations

### Advantages

* **Simplicity and Speed:** Logarithmic transformations allow us to fit complex curves quickly using standard linear matrix calculations, without needing slow iterative optimization loops.
* **Clear Insights:** Converting curves into straight lines makes it easy to visualize trends and calculate goodness-of-fit metrics like $R^2$.

### Limitations

* **Sensitivity to Low Values:** Logarithmic transformations distort the residual errors, making the model highly sensitive to small numbers near zero.
* **Zero and Negative Value Restrictions:** The equations fail completely if the dataset contains zero or negative numbers, as the logarithm of those values is undefined.



## 9. Conclusion

This study demonstrates how logarithmic transformations simplify non-linear regression analysis. By transforming power and exponential relationships into linear equations, we can use the reliable least-squares method to fit complex curves cleanly and efficiently. Choosing the right model comes down to how the data grows: use **Exponential Curves** for systems with rapid growth or decay proportional to their current size, and use **Power Curves** for physical or geometric properties that scale relative to size. This approach provides a fundamental tool for data analysis across engineering and scientific research.

# Numerical Differentiation using Finite Differences and Trapezoidal Rule for Numerical Integration



## 1. Objective

The objective of this report is to analyze, implement, and compare methods for approximating calculus operations numerically—specifically, **Numerical Differentiation** using finite difference approximations and **Numerical Integration** using the composite **Trapezoidal Rule**. This study focuses on how these methods approximate continuous derivatives and integrals from discrete data points, alongside their respective error behaviors.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

* **Numerical Differentiation:** Replaces continuous derivatives with algebraic combinations of discrete function values over a small step size $h$. It relies on truncating a local Taylor series expansion.
* **Numerical Integration (Trapezoidal Rule):** Approximates the definite integral $\int_{a}^{b} f(x) \, dx$ by dividing the area under the curve into $n$ equal sub-intervals. Instead of using rectangles, it connects the function values with straight lines, forming a series of trapezoids. The total integral is estimated by summing the areas of these trapezoids.

### Governing Formulae

#### Section A: Finite Difference Derivatives

Using an equally spaced step size $h$, the first derivative $f'(x)$ can be calculated using three main finite difference methods:


$$\text{Forward Difference: } f'(x) \approx \frac{f(x + h) - f(x)}{h}$$

$$\text{Backward Difference: } f'(x) \approx \frac{f(x) - f(x - h)}{h}$$

$$\text{Central Difference: } f'(x) \approx \frac{f(x + h) - f(x - h)}{2h}$$

To find the second derivative $f''(x)$, we use a symmetric central difference formula:


$$f''(x) \approx \frac{f(x + h) - 2f(x) + f(x - h)}{h^2}$$

#### Section B: Trapezoidal Integration Rule

Let an interval $[a, b]$ be partitioned into $n$ sub-intervals of equal width:


$$h = \frac{b - a}{n}$$


The grid points are defined as $x_i = a + ih$ for $i = 0, 1, 2, \dots, n$.

The area of a single trapezoid over sub-interval $[x_i, x_{i+1}]$ is:


$$\text{Area}_i = \frac{h}{2} \left[ f(x_i) + f(x_{i+1}) \right]$$

Summing all sub-intervals yields the **Composite Trapezoidal Rule Formula**:


$$\int_{a}^{b} f(x) \, dx \approx \frac{h}{2} \left[ f(x_0) + 2f(x_1) + 2f(x_2) + \dots + 2f(x_{n-1}) + f(x_n) \right]$$

$$\int_{a}^{b} f(x) \, dx \approx \frac{h}{2} \left[ y_0 + y_n + 2 \sum_{i=1}^{n-1} y_i \right]$$

### Error & Convergence Analysis

* **Differentiation Accuracy:** Forward and backward differences have first-order accuracy ($\mathcal{O}(h)$), while the central difference formulas provide higher accuracy with second-order scaling ($\mathcal{O}(h^2)$).
* **Integration Convergence:** The Composite Trapezoidal Rule is **second-order accurate** ($\mathcal{O}(h^2)$). The total global truncation error is bounded by:

$$E_{global} = -\frac{(b - a)h^2}{12} f''(\xi)$$



where $\xi$ is some unknown value within the domain $(a, b)$. This error formula shows that if you double the number of intervals $n$ (cutting the step size $h$ in half), the integration error drops by a factor of **four**.



## 3. Algorithmic Framework (Numbered Steps)

### Composite Trapezoidal Integration Algorithm

1. **Input:** Define the function $f(x)$ (or a table of coordinates), lower integration limit $a$, upper integration limit $b$, and the number of sub-intervals $n$.
2. **Step Computation:** Calculate the interval step size $h = \frac{b - a}{n}$.
3. **Initialize Accumulator:** Compute the boundary contribution sum: $\text{Sum} = f(a) + f(b)$.
4. **Loop Through Inner Nodes:** For $i = 1$ to $n-1$:
* Compute the intermediate coordinate node: $x_i = a + i \cdot h$.
* Add the weighted inner value to the running total: $\text{Sum} = \text{Sum} + 2 \cdot f(x_i)$.


5. **Finalize Integral Scale:** Multiply the accumulated sum by the front scale factor: $\text{FinalIntegral} = \frac{h}{2} \cdot \text{Sum}$.
6. **Output:** Return the calculated numerical integral value.




## 5. Mathematical Verification Case Study

We want to analyze the function $f(x) = x^3$ over the domain $[1, 3]$ to find both its derivative at a point and its total area.

### Case A: Central Derivative Verification

Approximate $f'(2)$ using a step size of $h = 0.1$. The exact derivative is $f'(x) = 3x^2$, which equals **$12.0$** at $x = 2$.

* $f(2 + 0.1) = f(2.1) = (2.1)^3 = 9.261$
* f(2 - 0.1) = f(1.9) = (1.9)^3 = 6.859

$$f'_{central}(2) \approx \frac{9.261 - 6.859}{2(0.1)} = \frac{2.402}{0.2} = \mathbf{12.01}$$


$$\text{Absolute Truncation Error} = |12.0 - 12.01| = 0.01$$





### Case B: Composite Trapezoidal Integration Verification

Evaluate $\int_{1}^{3} x^3 \, dx$ using $n = 4$ sub-intervals.
The exact analytical value is $\left[ \frac{x^4}{4} \right]_1^3 = \frac{81 - 1}{4} = \mathbf{20.0}$.

* **Step size width:** $h = \frac{3 - 1}{4} = 0.5$.
* **Generate Grid Nodes and Evaluations:**

| Node Index ($i$) | Coordinate ($x_i$) | Value $y_i = f(x_i) = x_i^3$ | Weight Factor | Product |
| --- | --- | --- | --- | --- |
| **0** | $1.0$ (Boundary $a$) | $1.000$ | 1 | $1.000$ |
| **1** | $1.5$ | $3.375$ | 2 | $6.750$ |

* **Substitute into the Composite Trapezoidal Rule:**

$$\int_{1}^{3} x^3 \, dx \approx \frac{0.5}{2} \left[ 1.000 + 27.000 + 2(3.375 + 8.000 + 15.625) \right]$$


$$\text{Integral} \approx 0.25 \left[ 28.000 + 2(27.000) \right] = 0.25 \left[ 28.000 + 54.000 \right]$$


$$\text{Integral} \approx 0.25 \cdot 82.000 = \mathbf{20.5}$$


$$\text{Absolute Integration Error} = |20.0 - 20.5| = 0.5$$





## 6. Observations & Computational Characteristics

### Geometric Profile

The Trapezoidal Rule uses straight lines to approximate the curve in each sub-interval. Because it forms flat segments across a curved function, it naturally introduces a slight bias:

* **Overestimation:** If a function is **concave up** ($f''(x) > 0$), the straight trapezoidal lines sit slightly above the actual curve, leading to an estimate that is higher than the true area.
* **Underestimation:** If a function is **concave down** ($f''(x) < 0$), the trapezoidal segments slice underneath the curve, underestimating the total area.

### Comparative Analysis

| Feature | Finite Difference Differentiation | Trapezoidal Integration Rule |
| --- | --- | --- |
| **Mathematical Operation** | Computes local slope (derivative) | Computes total area (integral) |
| **Grid Requirements** | Requires very close neighboring points ($h \to 0$) | Spans globally over an interval $[a, b]$ |
| **Error Scaling Performance** | $\mathcal{O}(h)$ or $\mathcal{O}(h^2)$ | $\mathcal{O}(h^2)$ (Second-order global scaling) |
| **Noise Vulnerability** | **Extremely high** (Small noise causes large errors) | **Very low** (Integration averages out random noise) |
| **Boundary Limitations** | Cannot use central formulas at outer boundaries | Functions perfectly across boundaries and edges |



## 7. Real-World Applications

* **Robotics & Inertial Navigation:** Integrating acceleration data over time to estimate velocity and position coordinates.
* **Structural Safety Testing:** Computing the total strain energy or structural load absorption by integrating stress-strain curves.
* **Signal Analysis & Power Grids:** Calculating total energy consumption (kilowatt-hours) by integrating fluctuating current and voltage telemetry streams over time.
* **Aerodynamics Optimization:** Estimating drag forces on wings by calculating pressure gradients across airfoil surface points.



## 8. Advantages & Limitations

### Numerical Differentiation

* **Advantages:** Straightforward to calculate from discrete data logs without requiring analytical functions.
* **Limitations:** Highly sensitive to noise. Small measurement errors or random noise can cause massive spikes in the calculated slopes, making raw differentiation unstable.

### Trapezoidal Integration Rule

* **Advantages:** Simple to implement and highly stable. Summing individual areas acts as an average, which naturally smooths out and reduces the impact of random noise in your data.
* **Limitations:** Limited to low-order linear segments. If a function has sharp bends or rapid curves, achieving high accuracy requires using very small intervals ($n$), which increases computing time.



## 9. Conclusion

This report highlights the fundamental operations of numerical calculus. Numerical differentiation tracks local variations to find rates of change, but it is highly sensitive to noise and requires careful step sizing. On the other hand, numerical integration balances out data noise by summing up area segments, providing a highly stable way to find total cumulative values. The composite Trapezoidal Rule offers a robust, second-order accurate framework ($\mathcal{O}(h^2)$) for uniform grids, making it an essential tool for data analysis and real-time processing systems across engineering disciplines.

# Implementation of Simpson’s 1/3 Rule and Simpson’s 3/8 Rule



## 1. Objective / Aim

The objective of this report is to analyze, implement, and compare **Simpson’s 1/3 Rule** and **Simpson’s 3/8 Rule** for numerical integration. This study examines how higher-order polynomial approximations (quadratic and cubic parabolas) improve the accuracy of definite integrals compared to linear methods like the Trapezoidal rule, while evaluating their structural grid requirements.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

While the Trapezoidal rule connects data points with straight lines, Simpson's methods approximate the underlying function using higher-degree polynomials via Newton-Cotes formulas:

* **Simpson’s 1/3 Rule:** Approximates the function across every two adjacent sub-intervals using a second-degree quadratic parabola passing through three consecutive points.
* **Simpson’s 3/8 Rule:** Approximates the function across every three adjacent sub-intervals using a third-degree cubic parabola passing through four consecutive points.

### Governing Formulae

#### Section A: Composite Simpson’s 1/3 Rule

Let the interval $[a, b]$ be divided into an **even number of sub-intervals $n$** (where $n \pmod 2 = 0$) with a uniform step size $h = \frac{b - a}{n}$.

Integrating a quadratic interpolating polynomial over a double interval $[x_i, x_{i+2}]$ yields the local area:


$$\int_{x_i}^{x_{i+2}} f(x) \, dx \approx \frac{h}{3} \left[ f(x_i) + 4f(x_{i+1}) + f(x_{i+2}) \right]$$

Summing all pairs across the entire domain produces the **Composite Simpson’s 1/3 Rule Formula**:


$$\int_{a}^{b} f(x) \, dx \approx \frac{h}{3} \left[ y_0 + y_n + 4(y_1 + y_3 + \dots + y_{n-1}) + 2(y_2 + y_4 + \dots + y_{n-2}) \right]$$

$$\int_{a}^{b} f(x) \, dx \approx \frac{h}{3} \left[ y_0 + y_n + 4 \sum_{\text{odd } i} y_i + 2 \sum_{\text{even } i} y_i \right]$$

#### Section B: Composite Simpson’s 3/8 Rule

Let the interval $[a, b]$ be divided into a **number of sub-intervals $n$ that is a multiple of three** (where $n \pmod 3 = 0$) with a uniform step size $h = \frac{b - a}{n}$.

Integrating a cubic interpolating polynomial over a triple interval $[x_i, x_{i+3}]$ yields the local area:


$$\int_{x_i}^{x_{i+3}} f(x) \, dx \approx \frac{3h}{8} \left[ f(x_i) + 3f(x_{i+1}) + 3f(x_{i+2}) + f(x_{i+3}) \right]$$

Summing all triplets across the entire domain produces the **Composite Simpson’s 3/8 Rule Formula**:


$$\int_{a}^{b} f(x) \, dx \approx \frac{3h}{8} \left[ y_0 + y_n + 3(y_1 + y_2 + y_4 + y_5 + \dots + y_{n-1}) + 2(y_3 + y_6 + \dots + y_{n-3}) \right]$$

$$\int_{a}^{b} f(x) \, dx \approx \frac{3h}{8} \left[ y_0 + y_n + 3 \sum_{i \neq \text{mult } 3} y_i + 2 \sum_{i = \text{mult } 3} y_i \right]$$

### Error & Convergence Analysis

* **Simpson's 1/3 Rule Error:** The global truncation error is bounded by:

$$E_{1/3} = -\frac{(b - a)h^4}{180} f^{(4)}(\xi)$$



The error scales with **fourth-order accuracy** ($\mathcal{O}(h^4)$). Surprisingly, although derived from a quadratic polynomial (which has a zero third derivative), the error term depends on the *fourth* derivative. This means Simpson's 1/3 rule integrates cubic functions exactly.
* **Simpson's 3/8 Rule Error:** The global truncation error is bounded by:

$$E_{3/8} = -\frac{(b - a)h^4}{80} f^{(4)}(\xi)$$



It also exhibits **fourth-order accuracy** ($\mathcal{O}(h^4)$). While slightly more precise locally than the 1/3 rule for a given step size $h$, it requires an extra function evaluation per block, making them comparable in global performance.



## 3. Algorithmic Framework (Numbered Steps)

### Composite Simpson’s 1/3 Rule Algorithm

1. **Input:** Define function $f(x)$, limits $a$ and $b$, and number of intervals $n$.
2. **Constraint Check:** Check if $n$ is even ($n \pmod 2 == 0$). If not, terminate with an execution error.
3. **Initialize Parameters:** Compute step size $h = \frac{b - a}{n}$ and set $\text{Sum} = f(a) + f(b)$.
4. **Loop Through Inner Nodes:** For $i = 1$ to $n-1$:
* If $i$ is odd, set $\text{Sum} = \text{Sum} + 4 \cdot f(a + i \cdot h)$.
* If $i$ is even, set $\text{Sum} = \text{Sum} + 2 \cdot f(a + i \cdot h)$.


5. **Finalize Value:** Calculate $\text{IntegralValue} = \frac{h}{3} \cdot \text{Sum}$ and output the result.

### Composite Simpson’s 3/8 Rule Algorithm

1. **Input:** Define function $f(x)$, limits $a$ and $b$, and number of intervals $n$.
2. **Constraint Check:** Check if $n$ is a multiple of 3 ($n \pmod 3 == 0$). If not, terminate with an error.
3. **Initialize Parameters:** Compute step size $h = \frac{b - a}{n}$ and set $\text{Sum} = f(a) + f(b)$.
4. **Loop Through Inner Nodes:** For $i = 1$ to $n-1$:
* If $i$ is a multiple of 3, set $\text{Sum} = \text{Sum} + 2 \cdot f(a + i \cdot h)$.
* Otherwise, set $\text{Sum} = \text{Sum} + 3 \cdot f(a + i \cdot h)$.


5. **Finalize Value:** Calculate $\text{IntegralValue} = \frac{3h}{8} \cdot \text{Sum}$ and output the result.

# Implementation of Modified Euler’s Method and Runge–Kutta Method (RK4)



## 1. Objective

The objective of this report is to analyze, implement, and compare two widely used numerical techniques for solving Ordinary Differential Equations (ODEs) with a given initial condition (Initial Value Problems, or IVPs): the **Modified Euler’s Method** (a predictor-corrector approach) and the **Classic Fourth-Order Runge–Kutta Method (RK4)**. This study focuses on their mathematical derivations, step-by-step algorithms, local/global error profiles, and practical stability bounds.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

Consider a first-order Ordinary Differential Equation of the form:


$$\frac{dy}{dx} = f(x, y), \quad \text{with initial condition } y(x_0) = y_0$$


Solving an IVP numerically involves starting at a known coordinate $(x_0, y_0)$ and marching forward in discrete steps of size $h$ ($x_{k+1} = x_k + h$) to estimate the next state $y_{k+1}$.

* **Modified Euler’s Method:** Standard Euler's method projects forward along a single tangent line evaluated at the beginning of the interval ($x_k$). Because the slope changes across the step, this introduces a noticeable drift. The Modified Euler method fixes this by using a **predictor-corrector framework**. It projects an initial guess (prediction) at the next point, evaluates the slope there, and then averages the starting and ending slopes to make a refined, balanced step (correction).
* **Runge–Kutta Method (RK4):** Instead of using just the endpoints, the RK4 method samples the derivative at **four strategic points** across the interval: once at the start, twice at midpoints, and once at the end. By taking a weighted average of these four slopes, it cancels out lower-order truncation errors, matching the high accuracy of a fourth-order Taylor series expansion without requiring complex analytical higher derivatives.

### Governing Formulae

#### Section A: Modified Euler’s Method (Heun's Predictor-Corrector)

For a step size $h$:

1. **Predictor Step ($\tilde{y}_{k+1}$):** Project an initial guess using the tangent at the start of the interval:

$$\tilde{y}_{k+1} = y_k + h \cdot f(x_k, y_k)$$


2. **Corrector Step ($y_{k+1}$):** Refine the estimate by averaging the starting slope and the predicted ending slope:

$$y_{k+1} = y_k + \frac{h}{2} \left[ f(x_k, y_k) + f(x_k + h, \tilde{y}_{k+1}) \right]$$



#### Section B: Fourth-Order Runge–Kutta Method (RK4)

For a step size $h$, the update step is calculated as:


$$y_{k+1} = y_k + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4)$$


where the four directional slopes are defined as:

* $k_1 = h \cdot f(x_k, y_k)$      *(Slope at the beginning of the interval)*
* $k_2 = h \cdot f\left(x_k + \frac{h}{2}, y_k + \frac{k_1}{2}\right)$      *(Slope at the midpoint, using $k_1$ to estimate $y$)*
* $k_3 = h \cdot f\left(x_k + \frac{h}{2}, y_k + \frac{k_2}{2}\right)$      *(Slope at the midpoint, using $k_2$ to estimate $y$)*
* $k_4 = h \cdot f(x_k + h, y_k + k_3)$      *(Slope at the end of the interval, using $k_3$ to estimate $y$)*

### Error & Convergence Analysis

* **Modified Euler’s Method:** The local truncation error per step scales as $\mathcal{O}(h^3)$. When accumulated over an integration interval, the **global truncation error is second-order accurate ($\mathcal{O}(h^2)$)**.
* **Runge–Kutta Method (RK4):** The local truncation error per step scales as $\mathcal{O}(h^5)$. Accumulated across the entire simulation domain, the **global truncation error is fourth-order accurate ($\mathcal{O}(h^4)$)**. Cutting the step size $h$ in half reduces the global error by a factor of **sixteen** ($2^4$), providing exceptional accuracy and high stability.



## 3. Algorithmic Framework (Numbered Steps)

### Modified Euler’s Method Algorithm

1. **Input:** Define derivative function $f(x, y)$, starting state variables $x_0$ and $y_0$, target point $x_{target}$, and step size $h$.
2. **Calculate Steps:** Compute total iteration cycles required: $N = \text{round}\left(\frac{x_{target} - x_0}{h}\right)$.
3. **Loop Updates:** For $k = 0$ to $N-1$:
* Compute starting slope: $m_1 = f(x_k, y_k)$.
* Predict intermediate state: $y_{pred} = y_k + h \cdot m_1$.
* Compute ending slope estimate: $m_2 = f(x_k + h, y_pred)$.
* Correct and update state: $y_{k+1} = y_k + \frac{h}{2}(m_1 + m_2)$.
* Advance independent variable: $x_{k+1} = x_k + h$.


4. **Output:** Return the final array coordinates tracking the solution curve.

### Classic Runge–Kutta Method (RK4) Algorithm

1. **Input:** Define function $f(x, y)$, initial states $x_0, y_0$, target point $x_{target}$, and step size $h$.
2. **Loop Updates:** While $x_k < x_{target}$:
* If $(x_k + h) > x_{target}$, truncate step size: $h = x_{target} - x_k$.
* Compute $k_1 = h \cdot f(x_k, y_k)$.
* Compute $k_2 = h \cdot f(x_k + 0.5h, y_k + 0.5k_1)$.
* Compute $k_3 = h \cdot f(x_k + 0.5h, y_k + 0.5k_2)$.
* Compute $k_4 = h \cdot f(x_k + h, y_k + k_3)$.
* Update dependent state: $y_{k+1} = y_k + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4)$.
* Update independent state: $x_{k+1} = x_k + h$.


3. **Output:** Return the final evaluated step value $y_{k+1}$.




## 5. Mathematical Verification Case Study

Let us solve the following first-order Initial Value Problem:


$$\frac{dy}{dx} = x + y, \quad \text{with initial condition } y(0) = 1$$


We want to estimate the value of $y$ at **$x = 0.2$** using a step size of **$h = 0.2$** (executing exactly one step).

The exact analytical solution derived via integrating factors is $y(x) = 2e^x - x - 1$.
At $x = 0.2$, the **exact value is $2e^{0.2} - 0.2 - 1 \approx 2(1.22140276) - 1.2 = \mathbf{1.24280552}$**.

### Case A: Modified Euler’s Method Verification

* **Evaluate starting conditions:** $x_0 = 0, \, y_0 = 1$. The starting slope is $f(0, 1) = 0 + 1 = 1.0$.
* **Predictor Step:**

$$\tilde{y}_1 = y_0 + h \cdot f(x_0, y_0) = 1 + (0.2)(1.0) = 1.2$$


* **Evaluate slope at predicted point:** $f(0.2, 1.2) = 0.2 + 1.2 = 1.4$.
* **Corrector Step:** Average the starting slope ($1.0$) and the predicted slope ($1.4$):

$$y_1 = y_0 + \frac{0.2}{2} \left[ f(0, 1) + f(0.2, 1.2) \right] = 1 + 0.1 \left[ 1.0 + 1.4 \right]$$


$$y_1 = 1 + 0.1(2.4) = \mathbf{1.24000000}$$


* **Error Measurement:** Absolute error is $|1.24280552 - 1.24000000| = \mathbf{2.81 \times 10^{-3}}$.



### Case B: Runge–Kutta Method (RK4) Verification

Using the same initial conditions $x_0 = 0, \, y_0 = 1$ and step size $h = 0.2$:

1. $$k_1 = h \cdot f(x_0, y_0) = 0.2 \cdot (0 + 1) = \mathbf{0.20000000}$$


2. $$k_2 = h \cdot f\left(x_0 + \frac{h}{2}, y_0 + \frac{k_1}{2}\right) = 0.2 \cdot f\left(0 + 0.1, 1 + 0.1\right) = 0.2 \cdot (0.1 + 1.1) = \mathbf{0.24000000}$$


3. $$k_3 = h \cdot f\left(x_0 + \frac{h}{2}, y_0 + \frac{k_2}{2}\right) = 0.2 \cdot f\left(0 + 0.1, 1 + 0.12\right) = 0.2 \cdot (0.1 + 1.12) = \mathbf{0.24400000}$$


4. $$k_4 = h \cdot f\left(x_0 + h, y_0 + k_3\right) = 0.2 \cdot f\left(0.2, 1 + 0.244\right) = 0.2 \cdot (0.2 + 1.244) = \mathbf{0.28880000}$$



Now, compute the weighted average of these four slopes to update $y_1$:


$$y_1 = y_0 + \frac{1}{6}(k_1 + 2k_2 + 2k_3 + k_4)$$

$$y_1 = 1 + \frac{1}{6}\left[ 0.2000 + 2(0.2400) + 2(0.2440) + 0.2888 \right]$$

$$y_1 = 1 + \frac{1}{6}\left[ 0.2000 + 0.4800 + 0.4880 + 0.2888 \right] = 1 + \frac{1.4568}{6}$$

$$y_1 = 1 + 0.24280000 = \mathbf{1.24280000}$$

* **Error Measurement:** Absolute error is $|1.24280552 - 1.24280000| = \mathbf{5.52 \times 10^{-6}}$.

In just a single step, the RK4 method achieves a remarkably tiny error, capturing five correct decimal places and showcasing its superior accuracy compared to the Modified Euler method.



## 6. Observations & Computational Characteristics

### Geometric Profile Analysis

* **Modified Euler** uses a straightforward predictor-corrector loop. While this prevents the solution from drifting as quickly as standard Euler's method, it relies entirely on linear trends at the boundaries, which causes errors to accumulate when solving highly non-linear or rapidly turning equations.
* **RK4** acts like an adaptive tracking system. By sampling slopes across the entire interval (at the start, midpoints, and end), it acts like an ensemble average that naturally smooths out sharp curves, providing a highly precise and stable path.

### Comparative Analysis

| Feature | Modified Euler’s Method | Runge–Kutta Method (RK4) |
| --- | --- | --- |
| **Method Category** | Predictor-Corrector Single-Step | Multi-Stage Single-Step Explicit |

| **Local Truncation Error** | $\mathcal{O}(h^3)$ | $\mathcal{O}(h^5)$ |
| **Global Truncation Error ($p$)** | $\mathcal{O}(h^2)$ (Second-order accurate) | $\mathcal{O}(h^4)$ (Fourth-order accurate) |

# Finite Difference Method for Heat Equation



## 1. Objective

The objective of this computational analysis is to analyze, implement, and evaluate the **Finite Difference Method (FDM)** for solving the one-dimensional transient **Heat Equation** (a parabolic Partial Differential Equation, or PDE). This study focuses on transforming continuous differential operators into discrete algebraic systems using the **Forward-Time Central-Space (FTCS)** explicit framework, while evaluating its structural grid configuration and numerical stability limits.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

The one-dimensional transient heat equation models how thermal energy diffuses through a homogeneous medium over time. The continuous governing parabolic PDE is expressed as:


$$\frac{\partial u}{\partial t} = \alpha \frac{\partial^2 u}{\partial x^2}$$


where:

* $u(x, t)$ represents the temperature distribution at spatial position $x$ and time $t$.
* $\alpha$ is the thermal diffusivity coefficient of the material ($\alpha > 0$).

To solve this equation across a spatial domain $x \in [0, L]$ and time $t \in [0, T]$, we discretize the domain into a structured mesh grid. Let $\Delta x$ be the spatial step size and $\Delta t$ be the temporal step size. The discrete grid coordinates are defined as:


$$x_i = i\Delta x \quad (i = 0, 1, \dots, N_x)$$

$$t_n = n\Delta t \quad (n = 0, 1, \dots, N_t)$$


The discrete temperature at node $i$ and time level $n$ is denoted as $u_i^n \approx u(x_i, t_n)$.

### Governing Formulae

Using Taylor series expansions, the continuous derivatives are replaced with algebraic finite difference approximations:

* **First-Order Forward Difference in Time:**

$$\frac{\partial u}{\partial t} \approx \frac{u_i^{n+1} - u_i^n}{\Delta t} + \mathcal{O}(\Delta t)$$


* **Second-Order Central Difference in Space:**

$$\frac{\partial^2 u}{\partial x^2} \approx \frac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{(\Delta x)^2} + \mathcal{O}((\Delta x)^2)$$



Substituting these discrete terms back into the governing PDE yields the **FTCS Explicit Scheme**:


$$\frac{u_i^{n+1} - u_i^n}{\Delta t} = \alpha \left[ \frac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{(\Delta x)^2} \right]$$

Isolating the unknown future value $u_i^{n+1}$ gives the explicit update relationship:


$$u_i^{n+1} = u_i^n + r \left( u_{i+1}^n - 2u_i^n + u_{i-1}^n \right)$$


where $r$ is the dimensionless **Mesh Fourier Number** (or diffusion metric), defined as:


$$r = \alpha \frac{\Delta t}{(\Delta x)^2}$$

Rearranging the terms yields the final computational formula:


$$u_i^{n+1} = r u_{i-1}^n + (1 - 2r) u_i^n + r u_{i+1}^n$$

### Stability & Convergence Criteria

Because the FTCS scheme is explicit, it calculates future states using only known values from the current time step. However, this convenience comes with a strict stability constraint known as the **Von Neumann Stability Criterion**.

For the solution to remain stable and prevent numerical errors from growing exponentially over time, the mesh Fourier number must satisfy:


$$r = \alpha \frac{\Delta t}{(\Delta x)^2} \le \frac{1}{2}$$

* **If $r \le 0.5$:** Errors decay smoothly, and the numerical solution safely converges to the exact physical solution as the grid is refined.
* **If $r > 0.5$:** The coefficients create an unphysical mathematical loop, triggering severe oscillations that quickly blow up the simulation.

### Order of Convergence

The truncation error for this explicit scheme is first-order accurate in time and second-order accurate in space, denoted as **$\mathcal{O}(\Delta t + (\Delta x)^2)$**.



## 3. Algorithmic Framework (Numbered Steps)

### FTCS Explicit Heat Solver Algorithm

1. **Input Parameters:** Define domain length $L$, total simulation time $T$, thermal diffusivity $\alpha$, spatial nodes $N_x$, and time steps $N_t$.
2. **Compute Grid Spacing:** Calculate $\Delta x = L / N_x$ and $\Delta t = T / N_t$.
3. **Stability Check:** Compute $r = \alpha \Delta t / (\Delta x)^2$. If $r > 0.5$, terminate the simulation with a critical warning: "Stability criterion violated ($r > 0.5$). Reduce $\Delta t$ or increase $\Delta x$."
4. **Initialize Arrays:** Construct a 1D temperature array `U` of size $N_x + 1$ to hold values for the current time step, and an array `U_next` of the same size for the next step.
5. **Apply Initial Conditions (IC):** Initialize the array elements `U[i]` at $t=0$ using the problem's initial temperature distribution function $u(x, 0) = f(x)$.
6. **Temporal Marching Loop:** For each time step $n = 0$ to $N_t - 1$:
* **Enforce Boundary Conditions (BC):** Set fixed values at the boundaries: `U_next[0] = Boundary_Left` and `U_next[N_x] = Boundary_Right`.
* **Internal Spatial Loop:** For each internal node $i = 1$ to $N_x - 1$, calculate the future temperature:
`U_next[i] = r * U[i-1] + (1.0 - 2.0*r) * U[i] + r * U[i+1]`
* **Update State:** Copy the values from `U_next` into `U` to prepare for the next iteration.


7. **Output:** Return or plot the final temperature distribution profile across the spatial domain.



## 5. Mathematical Verification Case Study

Consider a thin metal rod of length $L = 1.0\,\text{m}$ with a thermal diffusivity coefficient of $\alpha = 1.0\,\text{m}^2/\text{s}$.

* **Initial Condition:** The rod starts with a sinusoidal temperature profile: $u(x, 0) = \sin(\pi x)$.
* **Boundary Conditions:** Both ends are held fixed at freezing point: $u(0, t) = 0$ and $u(1, t) = 0$.

We choose to discretize the rod using $N_x = 4$ spatial intervals, giving a step size of $\Delta x = 1.0 / 4 = 0.25\,\text{m}$. We want to find the temperature distribution after exactly one time step ($\Delta t = 0.02\,\text{s}$).

### Step 1: Stability Verification

$$r = \alpha \frac{\Delta t}{(\Delta x)^2} = 1.0 \cdot \frac{0.02}{(0.25)^2} = \frac{0.02}{0.0625} = \mathbf{0.32}$$


Since $r = 0.32 \le 0.5$, the simulation is stable, allowing us to proceed safely.

### Step 2: Set Up Initial Profiles ($t = 0$)

Calculate the initial temperatures at each spatial grid node $i$:

* $u_0^0 = \sin(0) = 0$
* $u_1^0 = \sin(\pi \cdot 0.25) = \sin(45^\circ) \approx 0.7071$
* $u_2^0 = \sin(\pi \cdot 0.50) = \sin(90^\circ) = 1.0000$
* $u_3^0 = \sin(\pi \cdot 0.75) = \sin(135^\circ) \approx 0.7071$
* $u_4^0 = \sin(\pi \cdot 1.00) = \sin(180^\circ) = 0$

### Step 3: Compute the Explicit Time Step ($n = 0 \to 1$)

Apply the boundary conditions: $u_0^1 = 0$ and $u_4^1 = 0$. Now, calculate the updated temperatures for the internal nodes using $r = 0.32$ and $(1 - 2r) = 0.36$:

1. **Evaluate Node $i = 1$:**

$$u_1^1 = r u_0^0 + (1 - 2r) u_1^0 + r u_2^0$$


$$u_1^1 = 0.32(0) + 0.36(0.7071) + 0.32(1.0000) = 0 + 0.2546 + 0.3200 = \mathbf{0.5746}$$


2. **Evaluate Node $i = 2$:**

$$u_2^1 = r u_1^0 + (1 - 2r) u_2^0 + r u_3^0$$


$$u_2^1 = 0.32(0.7071) + 0.36(1.0000) + 0.32(0.7071) = 0.2263 + 0.3600 + 0.2263 = \mathbf{0.8126}$$

# Finite Difference Method for Wave Equation



## 1. Objective

The objective of this report is to analyze, implement, and evaluate the **Finite Difference Method (FDM)** for solving the one-dimensional **Wave Equation** (a hyperbolic Partial Differential Equation, or PDE). This study focuses on transforming continuous second-order differential operators into discrete algebraic systems using a central-difference time-stepping explicit framework, while evaluating its structural grid configuration and numerical stability limits.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

The one-dimensional wave equation models how displacements or disturbances (such as a vibrating string, acoustic pressure, or electromagnetic waves) propagate through a medium over time. The continuous governing hyperbolic PDE is expressed as:


$$\frac{\partial^2 u}{\partial t^2} = c^2 \frac{\partial^2 u}{\partial x^2}$$


where:

* $u(x, t)$ represents the wave displacement at spatial position $x$ and time $t$.
* $c$ is the constant propagation speed of the wave through the medium ($c > 0$).

To solve this equation across a spatial domain $x \in [0, L]$ and time $t \in [0, T]$, we discretize the domain into a structured mesh grid. Let $\Delta x$ be the spatial step size and $\Delta t$ be the temporal step size. The discrete grid coordinates are defined as:


$$x_i = i\Delta x \quad (i = 0, 1, \dots, N_x)$$

$$t_n = n\Delta t \quad (n = 0, 1, \dots, N_t)$$


The discrete displacement at node $i$ and time level $n$ is denoted as $u_i^n \approx u(x_i, t_n)$.

### Governing Formulae

Using Taylor series expansions, both second-order continuous derivatives are replaced with second-order central difference approximations:

* **Second-Order Central Difference in Time:**

$$\frac{\partial^2 u}{\partial t^2} \approx \frac{u_i^{n+1} - 2u_i^n + u_i^{n-1}}{(\Delta t)^2} + \mathcal{O}((\Delta t)^2)$$


* **Second-Order Central Difference in Space:**

$$\frac{\partial^2 u}{\partial x^2} \approx \frac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{(\Delta x)^2} + \mathcal{O}((\Delta x)^2)$$



Substituting these discrete terms back into the governing PDE yields the explicit central-difference scheme:


$$\frac{u_i^{n+1} - 2u_i^n + u_i^{n-1}}{(\Delta t)^2} = c^2 \left[ \frac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{(\Delta x)^2} \right]$$

Isolating the unknown future value $u_i^{n+1}$ gives the explicit update relationship:


$$u_i^{n+1} = 2u_i^n - u_i^{n-1} + \lambda^2 \left( u_{i+1}^n - 2u_i^n + u_{i-1}^n \right)$$


where $\lambda$ is the dimensionless **Courant Number** (or CFL parameter), defined as:


$$\lambda = c \frac{\Delta t}{\Delta x}$$

Rearranging terms yields the final computational formula for all standard time-stepping updates ($n \ge 1$):


$$u_i^{n+1} = \lambda^2 u_{i-1}^n + 2(1 - \lambda^2) u_i^n + \lambda^2 u_{i+1}^n - u_i^{n-1}$$

#### Special Handling for the First Time Step ($n = 0 \to 1$)

Because the wave equation is second-order in time, it requires two initial conditions: initial displacement $u(x,0) = f(x)$ and initial velocity $\frac{\partial u}{\partial t}(x,0) = g(x)$. The update formula requires a history value $u_i^{-1}$ at the first step, which does not exist.

Using a central difference for the initial velocity condition:


$$\frac{u_i^1 - u_i^{-1}}{2\Delta t} = g(x_i) \implies u_i^{-1} = u_i^1 - 2\Delta t \cdot g(x_i)$$

Substituting this into the main update equation for $n=0$ yields the specialized **first time-step formula**:


$$u_i^1 = \frac{\lambda^2}{2} u_{i-1}^0 + (1 - \lambda^2) u_i^0 + \frac{\lambda^2}{2} u_{i+1}^0 + \Delta t \cdot g(x_i)$$

### Stability & Convergence Criteria

Hyperbolic PDEs have a strict physical domain of dependence. For the numerical solution to remain stable and prevent errors from amplifying exponentially, the system must satisfy the **Courant-Friedrichs-Lewy (CFL) Condition**:


$$\lambda = c \frac{\Delta t}{\Delta x} \le 1$$

* **If $\lambda \le 1$:** The numerical domain of dependence encloses the physical domain, allowing information to propagate correctly. The scheme is stable.
* **If $\lambda > 1$:** The numerical scheme updates faster than the physical speed of the wave, leading to numerical divergence where the solution blows up completely.

### Order of Convergence

The scheme is second-order accurate in both time and space, denoted as **$\mathcal{O}((\Delta t)^2 + (\Delta x)^2)$**.



## 3. Algorithmic Framework (Numbered Steps)

### Explicit Wave Equation Solver Algorithm

1. **Input Parameters:** Define domain length $L$, simulation timeframe $T$, wave speed $c$, spatial nodes $N_x$, and time steps $N_t$.
2. **Compute Grid Spacing:** Calculate $\Delta x = L / N_x$ and $\Delta t = T / N_t$.
3. **CFL Stability Check:** Compute $\lambda = c \Delta t / \Delta x$. If $\lambda > 1.0$, terminate the program with an error: "CFL condition violated ($\lambda > 1.0$). Decrease $\Delta t$."
4. **Initialize Mesh Arrays:** Allocate three 1D arrays of size $N_x + 1$: `U_prev` ($n-1$), `U_curr` ($n$), and `U_next` ($n+1$).
5. **Apply Initial Profile ($n=0$):** Populate `U_curr[i]` using the initial displacement function $f(x)$.
6. **Execute Special First Step ($n=0 \to 1$):**
* Set boundary nodes `U_next[0]` and `U_next[N_x]` to 0 (assuming fixed boundaries).
* For inner nodes $i = 1$ to $N_x - 1$, apply the velocity-adjusted step formula to fill `U_next[i]`.
* Shift arrays: Copy `U_curr` to `U_prev`, and `U_next` to `U_curr`.


7. **Temporal Marching Loop:** For each subsequent time step $n = 1$ to $N_t - 1$:
* Enforce boundary states: `U_next[0] = 0` and `U_next[N_x] = 0`.
* Run internal spatial loop for $i = 1$ to $N_x - 1$:
`U_next[i] = lambda^2 * U_curr[i-1] + 2.0*(1.0 - lambda^2)*U_curr[i] + lambda^2 * U_curr[i+1] - U_prev[i]`
* Update array states: Copy `U_curr` to `U_prev`, then copy `U_next` to `U_curr`.


8. **Output:** Return or visualize the wave displacement profiles across the domain over time.

# Finite Difference Method for Laplace Equation



## 1. Objective

The objective of this report is to analyze, implement, and evaluate the **Finite Difference Method (FDM)** for solving the two-dimensional **Laplace Equation** (an elliptic Partial Differential Equation, or PDE). This study focuses on transforming continuous spatial differential operators into a discrete system of algebraic equations using the central-difference five-point stencil framework, and solving the resulting system iteratively using the **Gauss-Seidel Method**.



## 2. Mathematical Principle & Convergence Criteria

### Mathematical Background

The two-dimensional Laplace equation models steady-state boundary value problems where a physical quantity (such as temperature, electrostatic potential, or fluid stream function) has reached equilibrium. The continuous governing elliptic PDE is expressed as:


$$\nabla^2 u = \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0$$


where $u(x, y)$ represents the potential or temperature distribution across a two-dimensional spatial domain.

Unlike heat or wave equations, the Laplace equation does not depend on time. It is a pure **Boundary Value Problem (BVP)**. The solution inside the domain is entirely determined by the conditions enforced along the enclosing boundaries (**Dirichlet Boundary Conditions**).

To solve this equation across a rectangular domain $x \in [0, L]$ and $y \in [0, W]$, we discretize the space into a structured mesh grid. Let $\Delta x$ and $\Delta y$ be the spatial step sizes along the x and y axes, respectively. The discrete coordinates are:


$$x_i = i\Delta x \quad (i = 0, 1, \dots, N_x)$$

$$y_j = j\Delta y \quad (j = 0, 1, \dots, N_y)$$


The discrete potential at node $(i, j)$ is denoted as $u_{i,j} \approx u(x_i, y_j)$.

### Governing Formulae

Using Taylor series expansions, the second-order continuous spatial derivatives are replaced with second-order central difference approximations:


$$\frac{\partial^2 u}{\partial x^2} \approx \frac{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}{(\Delta x)^2} + \mathcal{O}((\Delta x)^2)$$

$$\frac{\partial^2 u}{\partial y^2} \approx \frac{u_{i,j+1} - 2u_{i,j} + u_{i,j-1}}{(\Delta y)^2} + \mathcal{O}((\Delta y)^2)$$

Substituting these terms into the Laplace equation yields the discrete algebraic governing system:


$$\frac{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}{(\Delta x)^2} + \frac{u_{i,j+1} - 2u_{i,j} + u_{i,j-1}}{(\Delta y)^2} = 0$$

Assuming a uniform square mesh grid where **$\Delta x = \Delta y = h$**, the denominators cancel out, simplifying the expression to:


$$u_{i+1,j} + u_{i-1,j} + u_{i,j+1} + u_{i,j-1} - 4u_{i,j} = 0$$

Isolating the central node potential $u_{i,j}$ yields the **Standard Five-Point Formula (Liebmann’s Equation)**:


$$u_{i,j} = \frac{1}{4} \left( u_{i+1,j} + u_{i-1,j} + u_{i,j+1} + u_{i,j-1} \right)$$

This formula states that at steady state, the value at any internal node is the exact arithmetic average of its four immediate neighbors.

### Algorithmic Solution via Gauss-Seidel Iteration

Because all internal nodes are coupled, they form a large simultaneous linear system ($A\mathbf{u} = \mathbf{b}$). Instead of invertng the matrix directly, we use the iterative **Gauss-Seidel Method**, which overwrites values in place as soon as they are computed:


$$u_{i,j}^{(k+1)} = \frac{1}{4} \left( u_{i+1,j}^{(k)} + u_{i-1,j}^{(k+1)} + u_{i,j+1}^{(k)} + u_{i,j-1}^{(k+1)} \right)$$


where $k$ represents the iteration counter.

### Stability & Convergence Criteria

Because elliptic equations are self-correcting and smooth out variations, iterative methods are **unconditionally stable**. The iterative loop is guaranteed to converge to a unique solution, provided that the system satisfies the **Scarborough Criterion** for diagonal dominance.

Convergence is achieved when the maximum absolute change between consecutive iterations falls below a specified error tolerance $\epsilon$:


$$\max_{i,j} \left| u_{i,j}^{(k+1)} - u_{i,j}^{(k)} \right| < \epsilon$$

### Order of Convergence

The spatial discretization error is second-order accurate across both dimensions, denoted as **$\mathcal{O}((h)^2)$**.


## 3. Algorithmic Framework (Numbered Steps)

### Iterative Finite Difference Laplace Solver Algorithm

1. **Input Parameters:** Define domain width $L$, height $W$, number of spatial subdivisions $N_x$ and $N_y$, absolute error tolerance $\epsilon$, and maximum allowable iterations $K_{max}$.
2. **Grid Setup:** Compute spatial step widths $\Delta x = L / N_x$ and $\Delta y = W / N_y$. Ensure $\Delta x = \Delta y = h$ for the standard five-point average formula.
3. **Initialize Mesh Matrix:** Allocate a 2D matrix `U` of size $(N_x + 1) \times (N_y + 1)$. Initialize all internal nodes to an initial guess (typically $0$ or the average of the boundaries).
4. **Enforce Dirichlet Boundary Conditions:** Populate the outer rows and columns of the matrix with fixed boundary values:
* Left Boundary: `U[0][j] = Bound_Left`
* Right Boundary: `U[N_x][j] = Bound_Right`
* Bottom Boundary: `U[i][0] = Bound_Bottom`
* Top Boundary: `U[i][N_y] = Bound_Top`


5. **Iterative Solver Loop:** For iteration $k = 1$ to $K_{max}$:
* Clear the error tracking variable: `max_diff = 0.0`.
* **Nested Spatial Loops:** For internal rows $j = 1$ to $N_y - 1$ and columns $i = 1$ to $N_x - 1$:
* Store the current value: `old_val = U[i][j]`.
* Compute the new value using the five-point average formula:
`U[i][j] = 0.25 * (U[i+1][j] + U[i-1][j] + U[i][j+1] + U[i][j-1])`
* Update the peak error indicator: `max_diff = max(max_diff, abs(U[i][j] - old_val))`.


* **Convergence Verification Check:** If `max_diff < ε`, terminate the loop early as convergence has been reached.


6. **Output:** Return the finalized steady-state potential matrix field `U`.




## 5. Mathematical Verification Case Study

Consider a square metallic plate of size $1.0\,\text{m} \times 1.0\,\text{m}$.

* **Boundary Conditions:** The top edge is maintained at a constant temperature of $100^\circ\text{C}$, while the other three edges are kept cold at $0^\circ\text{C}$.

We solve the internal nodes sequentially from bottom-left to top-right, immediately using the updated values as they become available:

$$u_{1,2}^{(1)} = \frac{1}{4} \left( u_{2,2}^{(0)} + u_{0,2} + u_{1,3} + u_{1,1}^{(1)} \right) = \frac{1}{4} \left( 0 + 0 + 100 + 0.0 \right) = \mathbf{25.0}$$


4. **Evaluate Node $(2,2)$:** Uses updated $u_{1,2}^{(1)} = 25.0$, $u_{2,1}^{(1)} = 0.0$, and boundary values $u_{3,2}=0, u_{2,3}=100$:

$$u_{2,2}^{(1)} = \frac{1}{4} \left( u_{3,2} + u_{1,2}^{(1)} + u_{2,3} + u_{2,1}^{(1)} \right) = \frac{1}{4} \left( 0 + 25.0 + 100 + 0.0 \right) = \frac{125}{4} = \mathbf{31.25}$$




### Step 3: Second Iteration Step ($k = 2$)

1. **Evaluate Node $(1,1)$:** Uses updated $u_{2,1}^{(1)}=0.0$ and $u_{1,2}^{(1)}=25.0$:

$$u_{1,1}^{(2)} = \frac{1}{4} \left( u_{2,1}^{(1)} + 0 + u_{1,2}^{(1)} + 0 \right) = \frac{1}{4} \left( 0 + 0 + 25.0 + 0 \right) = \mathbf{6.25}$$


2. **Evaluate Node $(2,1)$:** Uses updated $u_{1,1}^{(2)}=6.25$ and $u_{2,2}^{(1)}=31.25$:

$$u_{2,1}^{(2)} = \frac{1}{4} \left( 0 + u_{1,1}^{(2)} + u_{2,2}^{(1)} + 0 \right) = \frac{1}{4} \left( 0 + 6.25 + 31.25 + 0 \right) = \frac{37.5}{4} = \mathbf{9.375}$$


3. **Evaluate Node $(1,2)$:** Uses updated $u_{1,1}^{(2)}=6.25$ and $u_{2,2}^{(1)}=31.25$:

$$u_{1,2}^{(2)} = \frac{1}{4} \left( u_{2,2}^{(1)} + 0 + 100 + u_{1,1}^{(2)} \right) = \frac{1}{4} \left( 31.25 + 0 + 100 + 6.25 \right) = \frac{137.5}{4} = \mathbf{34.375}$$


4. **Evaluate Node $(2,2)$:** Uses updated $u_{1,2}^{(2)}=34.375$ and $u_{2,1}^{(2)}=9.375$:

$$u_{2,2}^{(2)} = \frac{1}{4} \left( 0 + u_{1,2}^{(2)} + 100 + u_{2,1}^{(2)} \right) = \frac{1}{4} \left( 0 + 34.375 + 100 + 9.375 \right) = \frac{143.75}{4} = \mathbf{35.9375}$$



As iterations continue ($k \to \infty$), the internal nodes stabilize at their exact steady-state values: $u_{1,1} = u_{2,1} = 11.11^\circ\text{C}$ and $u_{1,2} = u_{2,2} = 38.89^\circ\text{C}$. This demonstrates how the Gauss-Seidel method smoothly diffuses boundary conditions throughout the entire domain.




## 6. Observations & Computational Characteristics

### Stencil Geometry

The Laplace equation solver uses a symmetric five-point star configuration:

```
          (i, j+1) [Top]
             │
(i-1, j) ───(i, j)─── (i+1, j) [Right]
 [Left]      │
          (i, j-1) [Bottom]

```

At equilibrium, the potential at the central node $(i,j)$ is simply the average of its four orthogonal neighbors. This smooths out any sharp peaks or valleys inside the domain, ensuring that the maximum and minimum values always occur on the boundaries (**Maximum Principle**).

### Comparative Analysis (Iterative Linear Solvers)

| Feature | Jacobi Iteration Method | Gauss-Seidel Iteration Method | Successive Over-Relaxation (SOR) |
| --- | --- | --- | --- |
| **Data Memory** | Requires two matrices (reads from old, writes to new) | Requires a single matrix (overwrites values in place) | Requires a single matrix adjusted by a relaxation factor $\omega$ |
| **Update Rate Speed** | Slowest (Linear convergence) | Twice as fast as Jacobi | Fastest (Accelerated convergence via optimal $\omega$) |
| **Update Formula** | Uses only values from iteration $k$ | Uses new values from iteration $k+1$ as soon as available | Combines Gauss-Seidel with a weighted extrapolation step |
| **Implementation** | Simple but memory intensive | Simple and memory efficient | Requires tuning the optimal factor $\omega \in (1, 2)$ |

## 8. Advantages & Limitations

### Advantages

* **Guaranteed Convergence:** Elliptic equations are inherently self-correcting, making iterative routines stable and highly resistant to numerical divergence.
* **High Efficiency:** The Gauss-Seidel approach updates values in place, minimizing memory usage and making it ideal for processing large engineering simulation grids.

### Limitations

* **Slow Convergence on Large Grids:** While simple to implement, the method can slow down significantly as the grid size increases. This is because information can only travel one grid cell per iteration, requiring thousands of steps to diffuse boundary values across dense meshes.
* **Dirichlet Boundary Restrictions:** The basic five-point stencil assumes fixed boundary values. Handling curved geometries or insulated edges (**Neumann Boundary Conditions**) requires complex grid adjustments or ghost cells, which increases development overhead.
